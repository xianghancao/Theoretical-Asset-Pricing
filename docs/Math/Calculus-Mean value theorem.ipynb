# Mean value theorem

**均值定理 (Mean Value Theorem, MVT)** 是微积分中的一个核心定理，它描述了**函数在一段区间上的平均变化率**与其**瞬时变化率**之间的关系。

在没有特别说明的情况下，我们通常指的是**拉格朗日均值定理 (Lagrange's Mean Value Theorem)**。

---

## **均值定理的含义（几何和直观解释）**

均值定理的直观含义非常简洁且强大：

想象你有一条平滑（连续且可导）的曲线 $f(x)$，连接了两个点 $A$ 和 $B$。

1.  **割线的斜率 (平均变化率)：**
    连接 $A(a, f(a))$ 和 $B(b, f(b))$ 的直线段被称为**割线**。这条割线的斜率代表了函数在整个区间 $[a, b]$ 上的**平均变化率**。

2.  **切线的斜率 (瞬时变化率)：**
    函数在某一点 $c$ 处的导数 $f'(c)$ 代表了在该点切线的**瞬时变化率**。

**定理说的是：**
在连接 $A$ 和 $B$ 的这段曲线弧上，**至少存在一点 $c$**，使得函数在该点 $c$ 的**切线斜率**，恰好等于连接 $A$ 和 $B$ 的**割线斜率**。

用几何语言来说，就是：曲线上至少有一个点的切线，会与连接端点的割线**平行**。

---

## **数学表述 (拉格朗日均值定理)**

如果函数 $f(x)$ 满足以下两个条件：

1.  $f(x)$ 在**闭区间** $[a, b]$ 上**连续**。
2.  $f(x)$ 在**开区间** $(a, b)$ 内**可导**。

那么，在开区间 $(a, b)$ 内**至少存在一个点** $\boldsymbol{c}$，使得：

$$f'(c) = \frac{f(b) - f(a)}{b - a}$$

### **公式解读**

* 左边 $f'(c)$: 函数在 $c$ 点的**瞬时变化率**（切线斜率）。
* 右边 $\frac{f(b) - f(a)}{b - a}$: 函数在整个 $[a, b]$ 上的**平均变化率**（割线斜率）。

---

## **均值定理的种类**

均值定理家族主要有两大类：

### **1. 微分均值定理**

这是关于**导数**的定理，描述变化率。

* **罗尔定理 (Rolle's Theorem)：** 是均值定理的一个特例。如果 $f(a) = f(b)$（两端点值相等），那么 $f'(c) = 0$，即曲线上至少有一点的切线是**水平**的。
* **拉格朗日均值定理 (Lagrange's MVT)：** 上面介绍的标准形式。
* **柯西均值定理 (Cauchy's MVT)：** 是拉格朗日定理的推广，涉及两个函数 $f(x)$ 和 $g(x)$ 的关系。

### **2. 积分均值定理**

这是关于**积分**的定理，描述平均值。

* **积分第一均值定理：** 如果函数 $f(x)$ 在 $[a, b]$ 上连续，则在该区间内至少存在一点 $c$，使得：
    $$\int_a^b f(x) \,dx = f(c)(b-a)$$
    这表示函数在一个区间上的**积分（曲线下的面积）**，等于某一点的函数值 $f(c)$ 乘以区间长度 $(b-a)$。$f(c)$ 正是函数在 $[a, b]$ 上的**平均值**。
