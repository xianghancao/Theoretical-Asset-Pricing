# 罕见灾难模型 (Rare Disasters Model)



**罕见灾难模型 (Rare Disasters Model)** 由 Rietz (1988) 首次提出，后由 Barro (2006) 发展完善，是解决**股权溢价之谜 (Equity Premium Puzzle)** 和**低无风险利率之谜 (Risk-free Rate Puzzle)** 的又一重要理论模型。它通过引入低概率但高破坏性的**经济灾难 (Disasters)**，显著增加了资产的**系统性风险**，从而解释了资产定价的难题。

### 核心机制

简单的**消费资本资产定价模型 (CCAPM)** 假设消费增长服从对数正态分布，波动性较低。但在罕见灾难模型中，消费增长率的分布被修改，加入了**跳跃项 (Jump Term)**，代表经济灾难：

1.  **灾难的定义：** 灾难是指像大萧条、世界大战或重大金融危机那样的极端事件。这些事件的特征是**发生概率极低 (Rare)**，但一旦发生，会导致**消费和生产大幅度、永久性或半永久性地下降 (Large negative impact)**。
2.  **风险溢价来源：** 投资者非常厌恶这种极端的系统性风险。当灾难发生时，股票资产的价值会暴跌，同时，所有人的**消费也会遭受巨大冲击**。这种消费与股票收益在“坏状态”（灾难状态）下的**强负相关性**，使得股票的风险极高，投资者因此要求一个非常高的**风险溢价**来补偿这种风险。
3.  **高风险厌恶的必要性：** 为了让这种低概率事件对价格产生足够大的影响，模型必须假设投资者具有**高风险厌恶系数 ($\gamma$ 很高)**。在 Barro 的校准中，他发现只有将相对风险厌恶系数 $\gamma$ 设定在 **3.5 - 4** 左右（高于传统 CCAPM 的估计值 1-2），同时设定合理的灾难参数，才能成功匹配历史上的股权溢价（约 6%）和低无风险利率（约 1%）。

### 模型中的消费动态

在 Barro (2006) 的模型中，对数消费增长率 ($\Delta c_t$) 服从以下动态过程：

$$\Delta c_t = \mu + \varepsilon_t + J_t$$

其中：

* $\mu$: 平均消费增长率。
* $\varepsilon_t$: 正常的、高频率的、小幅的随机冲击（例如，服从 $N(0, \sigma^2)$）。
* $J_t$: **灾难跳跃项 (Disaster Jump)**。
    * $J_t$ 在绝大多数时间为 **0**。
    * $J_t$ 以一个极低的概率 $\pi$ 发生。
    * 一旦发生，它会是一个**巨大的负值**（例如，消费下降 25% 到 60%）。

**模型校准的关键参数（基于历史数据）：**

| 参数 | 符号 | Barro (2006) 估计值 | 含义 |
| :--- | :--- | :--- | :--- |
| 灾难发生概率（每年） | $\pi$ | $\approx 1.7\%$ | 每年发生一次灾难的概率。 |
| 灾难发生时的平均损失 | $b$ | $\approx 36\%$ | 灾难发生时，人均消费下降的比例。 |
| 相对风险厌恶系数 | $\gamma$ | $3.5 - 4$ | 投资者对风险的厌恶程度。 |

### 模型的成功与局限性

**成功之处：**

* **匹配股权溢价：** 灾难风险是系统性风险的重要组成部分，它使得具有高风险厌恶的投资者要求高额的溢价来持有股票。
* **匹配低无风险利率：** 在没有灾难的正常时期，灾难风险对无风险利率几乎没有影响，从而保持了低无风险利率。
* **简洁性：** 它在资产定价模型中引入了**尾部风险 (Tail Risk)**，这在金融实践中得到了广泛关注。

**局限性：**

* **数据依赖：** 灾难事件的数据样本极少，统计推断的稳健性较低。
* **参数要求：** 模型依赖于**高风险厌恶系数**，这可能与其他经济学观察（如风险投资决策）不一致。
* **难以解释资产收益的可预测性 (Predictability)：** 该模型主要是静态的，难以解释股票收益的**时变波动性**和**可预测性**，而这正是长久期风险模型等动态模型擅长的。
