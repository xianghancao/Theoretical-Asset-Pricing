## Idea

- 乘积数法则: $$\frac{d}{dt} \left( f(t) g(t) \right) = f'(t)g(t) + f(t)g'(t)$$
- 根据微积分基本定理，当对一个积分上限为变量的定积分求导时，直接将被积函数中的积分变量替换为积分上限即可。$g'(t) = \frac{d}{dt}\left(\int_0^t e^{\alpha u} c_u du\right)$，所以 $g'(t) = e^{\alpha t} c_t$。
- GBM的下限为0，而目前的consumption最低为$h_t$，所以，optimal consumption不可能是GBM
- explicit solution of GBM
- $dz \sim N(0, dt)$ means $z_s - z_t \sim N(0, s-t)$

## Solution

### **问题 (a) 的证明**

我们需要证明 $dh_t = (\beta c_t - \alpha h_t) dt$。

首先，我们有一个关于习惯 $h_t$ 的定义式：
$h_t = h_0e^{-\alpha t} + \beta \int_0^t e^{-\alpha(t-u)} c_u du$

我们可以将积分项中的 $e^{-\alpha t}$ 提到积分号外面，得到：
$h_t = h_0e^{-\alpha t} + \beta e^{-\alpha t} \int_0^t e^{\alpha u} c_u du$

接下来，我们对这个表达式关于时间 $t$ 求导，即计算 $\frac{dh_t}{dt}$。

1.  **对第一项 $h_0e^{-\alpha t}$ 求导：**
    $\frac{d}{dt}(h_0e^{-\alpha t}) = -\alpha h_0e^{-\alpha t}$

2.  **对第二项 $\beta e^{-\alpha t} \int_0^t e^{\alpha u} c_u du$ 求导：**
    这是一个乘积的求导，我们需要使用乘积法则：$\frac{d}{dt}(f(t)g(t)) = f'(t)g(t) + f(t)g'(t)$。
    这里 $f(t) = \beta e^{-\alpha t}$， $g(t) = \int_0^t e^{\alpha u} c_u du$。

    * $f'(t) = \frac{d}{dt}(\beta e^{-\alpha t}) = -\alpha \beta e^{-\alpha t}$
    * $g'(t) = \frac{d}{dt}\left(\int_0^t e^{\alpha u} c_u du\right)$。根据微积分基本定理，当对一个积分上限为变量的定积分求导时，直接将被积函数中的积分变量替换为积分上限即可。所以 $g'(t) = e^{\alpha t} c_t$。

    因此，第二项的导数为：
    $$(-\alpha \beta e^{-\alpha t})\left(\int_0^t e^{\alpha u} c_u du\right) + (\beta e^{-\alpha t})(e^{\alpha t} c_t)$$   $$= -\alpha \left(\beta e^{-\alpha t}\int_0^t e^{\alpha u} c_u du\right) + \beta c_t$$

现在，我们将两项的导数相加，得到 $\frac{dh_t}{dt}$：
$$\frac{dh_t}{dt} = -\alpha h_0e^{-\alpha t} - \alpha \left(\beta e^{-\alpha t}\int_0^t e^{\alpha u} c_u du\right) + \beta c_t$$

我们可以观察到，前两项可以合并。
$$-\alpha \left(h_0e^{-\alpha t} + \beta e^{-\alpha t}\int_0^t e^{\alpha u} c_u du\right) + \beta c_t$$

而括号内的表达式正好就是 $h_t$ 的定义式！
$$-\alpha h_t + \beta c_t$$

所以，我们最终得到：
$$\frac{dh_t}{dt} = \beta c_t - \alpha h_t$$

最后，将 $dt$ 乘到等式左边，就得到了我们要证明的表达式：
$$dh_t = (\beta c_t - \alpha h_t) dt$$

这表明，习惯 $h_t$ 的变化率 $dh_t$ 取决于两个因素：当前的**消费水平 $c_t$**（正向影响，系数为 $\beta$）和**当前的习惯存量 $h_t$**（负向影响，系数为 $-\alpha$）。这符合**习惯形成**的直观理解：消费会增加习惯，而习惯本身会随着时间衰减。

这是一个关于连续时间、完备市场经济中具有习惯形成偏好的个体期望效用最大化的问题。

**(a) 证明 $dh_t = (\beta c_t - \alpha h_t) \, dt$。$\alpha$ 和 $\beta$ 满足什么条件可以确保当期消费等于习惯水平时，习惯水平下降？**

**证明 $dh_t$ 的表达式：**

习惯水平 $h_t$ 可以写成如下形式：
$$
h_t = e^{-\alpha t} \left( h_0 + \beta \int_0^t e^{\alpha u} c_u \, du \right)
$$
对 $h_t$ 关于时间 $t$ 求导（使用莱布尼茨积分法则）：
$$
\frac{dh_t}{dt} = \frac{d}{dt} \left[ e^{-\alpha t} \left( h_0 + \beta \int_0^t e^{\alpha u} c_u \, du \right) \right]
$$
使用乘积法则 $\frac{d}{dt} [f(t)g(t)] = f'(t)g(t) + f(t)g'(t)$：
$$
\frac{dh_t}{dt} = (-\alpha e^{-\alpha t}) \left( h_0 + \beta \int_0^t e^{\alpha u} c_u \, du \right) + e^{-\alpha t} \left[ \frac{d}{dt} \left( h_0 + \beta \int_0^t e^{\alpha u} c_u \, du \right) \right]
$$
注意大括号内的第一项是 $h_t$，而第二项中积分的导数是 $\beta$ 乘以被积函数在上限 $t$ 处的值，即 $\beta e^{\alpha t} c_t$：
$$
\frac{dh_t}{dt} = -\alpha h_t + e^{-\alpha t} (\beta e^{\alpha t} c_t)
$$
$$
\frac{dh_t}{dt} = -\alpha h_t + \beta c_t
$$
因此，微分形式为：
$$
dh_t = (\beta c_t - \alpha h_t) \, dt
$$
**当期消费等于习惯水平时习惯水平下降的条件：**

当 $c_t = h_t$ 时，我们看 $dh_t/dt$ 的符号：
$$
\frac{dh_t}{dt} \Big|_{c_t = h_t} = \beta c_t - \alpha h_t = \beta h_t - \alpha h_t = (\beta - \alpha) h_t
$$
为了确保习惯水平下降，我们需要 $\frac{dh_t}{dt} < 0$。

假设习惯水平 $h_t$ 是非负的（$h_t \ge 0$），这是一个合理的经济假设。

1.  如果 $h_t > 0$，则需要 $\beta - \alpha < 0$，即 **$\beta < \alpha$**。
2.  如果 $h_t = 0$，则 $dh_t/dt = 0$，习惯水平不变。

因此，确保当 $c_t = h_t > 0$ 时习惯水平下降的条件是 **$\beta < \alpha$**。这表明习惯的消退率 ($\alpha$) 必须大于习惯的形成率 ($\beta$ )。

**(b) 在这些假设下，最优消费是否遵循几何布朗运动？**

在连续时间、完备市场的效用最大化问题中，最优消费率 $c_t$ 通常满足所谓的**“一阶条件”**：
$$
\zeta_t u_c(c_t, h_t) = K
$$
其中 $u_c(c_t, h_t) = \frac{\partial u}{\partial c}(c_t, h_t)$ 是边际效用，$K$ 是一个常数（依赖于初始财富），$\zeta_t$ 是**状态价格折现因子 (State-Price Deflator)**。

题目中给出的瞬时效用函数为 $u(c, h) = \frac{1}{1-\gamma} (c-h)^{1-\gamma}$。
边际效用为：
$$
u_c(c_t, h_t) = \frac{\partial}{\partial c_t} \left[ \frac{1}{1-\gamma} (c_t-h_t)^{1-\gamma} \right] = (c_t-h_t)^{-\gamma}
$$
令 $s_t = c_t - h_t$ 为**“剩余消费率”**（或“超额消费率”），则边际效用为 $u_c(c_t, h_t) = s_t^{-\gamma}$。

一阶条件变为：
$$
\zeta_t s_t^{-\gamma} = K \quad \Rightarrow \quad s_t^{-\gamma} = K \zeta_t^{-1} \quad \Rightarrow \quad s_t = K^{-1/\gamma} \zeta_t^{-1/\gamma}
$$
如果 **状态价格折现因子 $\zeta_t$ 遵循几何布朗运动 (Geometric Brownian Motion, GBM)**，那么剩余消费率 $s_t$（$\zeta_t$ 的幂次）也将遵循几何布朗运动。

**如果 $s_t$ 遵循几何布朗运动，那么 $c_t = s_t + h_t$ 是否遵循几何布朗运动？**

* $s_t$ 是一个伊藤过程（如果它是 GBM）。
* $h_t$ 是一个由 $c_t$ 决定的（非随机）过程（见 (a) 部分，如果 $c_t$ 是确定性的）。然而，在连续时间模型中，$c_t$ 通常是随机的，所以 $h_t$ 也是随机的（根据其积分形式或 SDE $dh_t = (\beta c_t - \alpha h_t) \, dt$）。如果 $c_t$ 是一个伊藤过程，$h_t$ 也是一个伊藤过程。
* $c_t$ 是两个伊藤过程 $s_t$ 和 $h_t$ 的**和**。**两个伊藤过程的和通常仍是一个伊藤过程，但它不是几何布朗运动**（除非其中一个或两个过程是退化的）。几何布朗运动的特征是其**对数**是布朗运动，这意味着它是一个乘法过程，而不是一个加法过程。

因此，**最优消费 $c_t$ 通常不遵循几何布朗运动**。

但是，题目在 (c) 部分假设**超额消费率 $\hat{c}_t = c_t - h_t$ 遵循几何布朗运动**。
注意到 $\hat{c}_t = c_t - h_t = s_t$。
**如果**（如 (c) 部分所假设的）$\hat{c}_t = s_t$ 是一个几何布朗运动，那么由一阶条件 $s_t = K^{-1/\gamma} \zeta_t^{-1/\gamma}$ 可知，状态价格折现因子 $\zeta_t$ 必须是几何布朗运动（或者 $\zeta_t^{-1/\gamma}$ 是 GBM，这等价于 $\zeta_t$ 是 GBM）。

因此，更准确的回答是：
* **消费 $c_t$ 通常不遵循几何布朗运动。**
* **超额消费率 $s_t = c_t - h_t$ 如果状态价格折现因子 $\zeta_t$ 遵循几何布朗运动，则它遵循几何布朗运动**。题目在 (c) 中接受了 $\hat{c}_t$ 遵循几何布朗运动这一假设。

**(c) 证明状态价格折现因子 $\zeta_t$ 具有 $\zeta_t = f(t) e^{-\delta t} \hat{c}_t^{-\gamma}$ 的形式。**

在完备市场中，状态价格折现因子 $\zeta_t$ 通常由边际效用给出：
$$
\zeta_t = e^{-\delta t} \frac{u_c(c_t, h_t)}{u_c(c_0, h_0)} \cdot k'
$$
其中 $k'$ 是一个归一化常数，通常取 $\frac{1}{u_c(c_0, h_0)}$。然而，题目中给出的 $\zeta_t$ 表达式更复杂，因为它必须同时满足两个条件：
1.  它是定价内核（核），即 $\zeta_t = e^{-\delta t} \frac{u_c(c_t, h_t)}{\text{常数}}$。
2.  它必须满足其定义的**期望形式**。

题目中给出了 $\zeta_t$ 的定义（它似乎是基于一个跨期边际效用）：
$$
\zeta_t = k e^{-\delta t} \left\{ u_c(c_t, h_t) + \beta E \left[ \int_t^T e^{-(\delta+\alpha)(s-t)} u_h(c_s, h_s) \, ds \Bigg| \mathcal{F}_t \right] \right\}
$$
其中 $k$ 是一个常数。
我们有 $u(c, h) = \frac{1}{1-\gamma} (c-h)^{1-\gamma}$。
超额消费率 $\hat{c}_t = c_t - h_t$。

**边际效用：**
$$
u_c(c_t, h_t) = (c_t - h_t)^{-\gamma} = \hat{c}_t^{-\gamma}
$$
$$
u_h(c_t, h_t) = \frac{\partial u}{\partial h_t} = \frac{1}{1-\gamma} (1-\gamma) (c_t-h_t)^{1-\gamma-1} \cdot (-1) = - (c_t-h_t)^{-\gamma} = - \hat{c}_t^{-\gamma}
$$

将这些代入 $\zeta_t$ 的定义式中：
$$
\zeta_t = k e^{-\delta t} \left\{ \hat{c}_t^{-\gamma} + \beta E \left[ \int_t^T e^{-(\delta+\alpha)(s-t)} (- \hat{c}_s^{-\gamma}) \, ds \Bigg| \mathcal{F}_t \right] \right\}
$$
$$
\zeta_t = k e^{-\delta t} \hat{c}_t^{-\gamma} \left\{ 1 - \beta E \left[ \int_t^T e^{-(\delta+\alpha)(s-t)} \frac{\hat{c}_s^{-\gamma}}{\hat{c}_t^{-\gamma}} \, ds \Bigg| \mathcal{F}_t \right] \right\}
$$
$$
\zeta_t = k e^{-\delta t} \hat{c}_t^{-\gamma} \left\{ 1 - \beta \int_t^T e^{-(\delta+\alpha)(s-t)} E \left[ \left( \frac{\hat{c}_s}{\hat{c}_t} \right)^{-\gamma} \Bigg| \mathcal{F}_t \right] \, ds \right\}
$$
题目假设 $\hat{c}_t = c_t - h_t$ 遵循几何布朗运动 (GBM)：
$$
d\hat{c}_t = \hat{c}_t [\hat{\mu} \, dt + \hat{\sigma} \, dz_t]
$$
注意 $\hat{\mu}$ 和 $\hat{\sigma}$ 必须与题目中给出的 $\tilde{\mu}$ 和 $\tilde{\sigma}$ 相符，因为 $\tilde{\mu}$ 和 $\tilde{\sigma}$ 是在 $Q$ 测度下的漂移和扩散系数（在 $Q$ 测度下， $\hat{c}_t$ 的漂移项是 $\tilde{\mu}$）。
$$
d\hat{c}_t = \hat{c}_t [\tilde{\mu} \, dt + \tilde{\sigma} \, dz_t]
$$
其中 $z$ 是 $Q$ 测度下的标准布朗运动。

对于一个 GBM $dX_t/X_t = \mu \, dt + \sigma \, dz_t$，$\ln X_t$ 是一个布朗运动 $d \ln X_t = (\mu - \sigma^2/2) \, dt + \sigma \, dz_t$。
对数消费比率 $\ln(\hat{c}_s/\hat{c}_t)$ 在 $Q$ 测度下是服从正态分布的：
$$
\ln \left( \frac{\hat{c}_s}{\hat{c}_t} \right) \sim N \left( \left( \tilde{\mu} - \frac{1}{2}\tilde{\sigma}^2 \right) (s-t), \tilde{\sigma}^2 (s-t) \right)
$$
我们计算期望 $E_t \left[ (\hat{c}_s/\hat{c}_t)^{-\gamma} \right]$。
令 $X = \ln (\hat{c}_s/\hat{c}_t)$, $X \sim N(m, v^2)$，其中 $m = (\tilde{\mu} - \frac{1}{2}\tilde{\sigma}^2) (s-t)$ 且 $v^2 = \tilde{\sigma}^2 (s-t)$。
我们要求 $E_t [e^{-\gamma X}]$。
对于正态分布 $X \sim N(m, v^2)$，我们有 $E[e^{\theta X}] = e^{\theta m + \frac{1}{2}\theta^2 v^2}$。
令 $\theta = -\gamma$：
$$
E_t \left[ \left( \frac{\hat{c}_s}{\hat{c}_t} \right)^{-\gamma} \right] = E_t [e^{-\gamma X}] = e^{-\gamma m + \frac{1}{2}(-\gamma)^2 v^2}
$$
代入 $m$ 和 $v^2$：
\begin{align*} -\gamma m + \frac{1}{2}\gamma^2 v^2 &= -\gamma \left( \tilde{\mu} - \frac{1}{2}\tilde{\sigma}^2 \right) (s-t) + \frac{1}{2}\gamma^2 \tilde{\sigma}^2 (s-t) \\ &= \left[ -\gamma \tilde{\mu} + \frac{1}{2}\gamma \tilde{\sigma}^2 + \frac{1}{2}\gamma^2 \tilde{\sigma}^2 \right] (s-t) \\ &= \left[ -\gamma \tilde{\mu} + \frac{1}{2}\tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t) \\ &= - \left[ \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t) \end{align*}
所以：
$$
E_t \left[ \left( \frac{\hat{c}_s}{\hat{c}_t} \right)^{-\gamma} \right] = e^{- \left[ \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t)}
$$
将此代回 $\zeta_t$ 的表达式：
$$
\zeta_t = k e^{-\delta t} \hat{c}_t^{-\gamma} \left\{ 1 - \beta \int_t^T e^{-(\delta+\alpha)(s-t)} e^{- \left[ \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t)} \, ds \right\}
$$
定义一个新参数 $\Theta$:
$$
\Theta = \delta + \alpha + \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma)
$$
注意题目中 $\Theta$ 的定义是 $\theta = \delta + \alpha + \tilde{\gamma} - \frac{1}{2} \tilde{\sigma} (1+\gamma^2)$。这与推导出的 $\Theta$ 形式略有不同。我们使用 $\tilde{\mu}$ 和 $\tilde{\sigma}$ 来匹配题目中的符号，并假设 $\tilde{\gamma}$ 就是 $\gamma \tilde{\mu}$。

使用题目中给出的 $\theta$ 定义（假设 $\tilde{\mu}$ 就是 $ \tilde{\mu}$ 而非 $\tilde{\gamma}$，因为 $\tilde{\mu}$ 是漂移项，$\tilde{\sigma}$ 是扩散项）：
$$
\theta = \delta + \alpha + \tilde{\mu} \gamma - \frac{1}{2} \tilde{\sigma} (1+\gamma^2)
$$
我们推导出的指数项是：
$$
(\delta+\alpha)(s-t) + \left[ \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t) = \left[ \delta + \alpha + \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t)
$$
注意到题目中的 $\theta$ 表达式似乎有**印刷错误**。它很可能应该包含 $\tilde{\mu}$ 和 $\tilde{\sigma}^2$，而不是 $\tilde{\sigma}$ 和 $1+\gamma^2$。

**假设题目给出的 $\theta$ 定义是正确的（我们必须使用它来匹配最终形式）：**
$$
\theta = \delta + \alpha + \tilde{\mu} \gamma - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma)
$$
（这里我们猜测题目中的 $\tilde{\gamma}$ 应该是 $\gamma \tilde{\mu}$，题目中的 $\frac{1}{2} \tilde{\sigma} (1+\gamma^2)$ 应该是 $\frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma)$。这是一个著名的习惯形成模型的结果，如 *Campbell and Cochrane (1999)* 或其变体。）

**假设题目给出的 $\theta$ 定义是正确的，并将其与我们推导的指数项** $\left[ \delta + \alpha + \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma) \right] (s-t)$ **匹配，则**
$$
\theta = \delta + \alpha + \gamma \tilde{\mu} - \frac{1}{2} \tilde{\sigma}^2 \gamma (1+\gamma)
$$

则指数部分为 $\theta (s-t)$。
$$
\zeta_t = k e^{-\delta t} \hat{c}_t^{-\gamma} \left\{ 1 - \beta \int_t^T e^{-\theta (s-t)} \, ds \right\}
$$
计算积分：令 $u = s-t$，则 $du = ds$。当 $s=t$ 时 $u=0$，当 $s=T$ 时 $u=T-t$。
$$
\int_t^T e^{-\theta (s-t)} \, ds = \int_0^{T-t} e^{-\theta u} \, du = \left[ - \frac{1}{\theta} e^{-\theta u} \right]_0^{T-t} = - \frac{1}{\theta} \left( e^{-\theta (T-t)} - e^0 \right) = \frac{1}{\theta} \left( 1 - e^{-\theta (T-t)} \right)
$$
代回 $\zeta_t$：
$$
\zeta_t = k e^{-\delta t} \hat{c}_t^{-\gamma} \left\{ 1 - \beta \frac{1}{\theta} \left( 1 - e^{-\theta (T-t)} \right) \right\}
$$
$$
\zeta_t = k \left[ 1 - \frac{\beta}{\theta} \left( 1 - e^{-\theta (T-t)} \right) \right] e^{-\delta t} \hat{c}_t^{-\gamma}
$$
定义确定性函数 $f(t)$：
$$
f(t) = k \left[ 1 - \frac{\beta}{\theta} \left( 1 - e^{-\theta (T-t)} \right) \right]
$$
这与题目中给出的 $f(t)$ 形式：
$$
f(t) = k \left( 1 - \frac{\beta}{\theta} (1 - e^{-\theta(T-t)}) \right)
$$
是**一致的**。

因此，
$$
\zeta_t = f(t) e^{-\delta t} \hat{c}_t^{-\gamma}
$$
证明完成。

**(d) 找到均衡短期利率和市场风险价格的表达式。与没有习惯的“简单模型”、CRRA 效用和遵循几何布朗运动的消费进行比较。**

我们有状态价格折现因子 $\zeta_t$ 的形式：
$$
\zeta_t = f(t) e^{-\delta t} \hat{c}_t^{-\gamma}
$$
其中 $f(t)$ 是确定性函数，$e^{-\delta t}$ 是时间折扣项，$\hat{c}_t$ 遵循 GBM：
$$
\frac{d\hat{c}_t}{\hat{c}_t} = \tilde{\mu} \, dt + \tilde{\sigma} \, dz_t
$$
其中 $z$ 是 $Q$ 测度下的标准布朗运动。

**短期利率 $r_t$：**

短期利率 $r_t$ 是状态价格折现因子 $\zeta_t$ 演化方程中 $dt$ 项的负漂移系数。一般而言，如果 $\zeta_t$ 是一个伊藤过程：
$$
\frac{d\zeta_t}{\zeta_t} = -r_t \, dt - \lambda_t \, dz_t
$$
其中 $r_t$ 是短期利率，$\lambda_t$ 是市场风险价格（在 $Q$ 测度下的 $\tilde{\mu}$）。

首先，我们需要找到 $\zeta_t$ 的伊藤微分。使用伊藤引理。
$\ln \zeta_t = \ln f(t) - \delta t - \gamma \ln \hat{c}_t$。
令 $g(\hat{c}_t, t) = \ln \zeta_t$。
$$
\frac{\partial g}{\partial t} = \frac{f'(t)}{f(t)} - \delta - \gamma \frac{\partial \ln \hat{c}_t}{\partial t} = \frac{f'(t)}{f(t)} - \delta
$$
（因为 $\ln \hat{c}_t$ 仅依赖于 $\hat{c}_t$，$\frac{\partial \ln \hat{c}_t}{\partial t}=0$）
$$
\frac{\partial g}{\partial \hat{c}_t} = - \gamma \frac{1}{\hat{c}_t}
$$
$$
\frac{\partial^2 g}{\partial \hat{c}_t^2} = - \gamma \left( - \frac{1}{\hat{c}_t^2} \right) = \frac{\gamma}{\hat{c}_t^2}
$$
$\hat{c}_t$ 的微分：
$$
d\hat{c}_t = \hat{c}_t \tilde{\mu} \, dt + \hat{c}_t \tilde{\sigma} \, dz_t
$$
应用伊藤引理 $dg = \frac{\partial g}{\partial t} dt + \frac{\partial g}{\partial \hat{c}_t} d\hat{c}_t + \frac{1}{2} \frac{\partial^2 g}{\partial \hat{c}_t^2} (d\hat{c}_t)^2$：
\begin{align*} d \ln \zeta_t &= \left( \frac{f'(t)}{f(t)} - \delta \right) dt + \left( - \frac{\gamma}{\hat{c}_t} \right) (\hat{c}_t \tilde{\mu} \, dt + \hat{c}_t \tilde{\sigma} \, dz_t) + \frac{1}{2} \left( \frac{\gamma}{\hat{c}_t^2} \right) (\hat{c}_t \tilde{\sigma})^2 \, dt \\ &= \left( \frac{f'(t)}{f(t)} - \delta \right) dt - \gamma \tilde{\mu} \, dt - \gamma \tilde{\sigma} \, dz_t + \frac{1}{2} \gamma \tilde{\sigma}^2 \, dt \\ &= \left( \frac{f'(t)}{f(t)} - \delta - \gamma \tilde{\mu} + \frac{1}{2} \gamma \tilde{\sigma}^2 \right) dt - \gamma \tilde{\sigma} \, dz_t \end{align*}
由于 $d \ln \zeta_t = \frac{d\zeta_t}{\zeta_t} - \frac{1}{2} (\frac{d\zeta_t}{\zeta_t})^2$，并且 $\frac{d\zeta_t}{\zeta_t} = \mu_{\zeta} \, dt + \sigma_{\zeta} \, dz_t$，
$d \ln \zeta_t = (\mu_{\zeta} - \frac{1}{2} \sigma_{\zeta}^2) \, dt + \sigma_{\zeta} \, dz_t$。
同时， $\sigma_{\zeta} = - \gamma \tilde{\sigma}$，所以 $\frac{1}{2} \sigma_{\zeta}^2 = \frac{1}{2} \gamma^2 \tilde{\sigma}^2$。

因此：
$$
\mu_{\zeta} - \frac{1}{2} \gamma^2 \tilde{\sigma}^2 = \frac{f'(t)}{f(t)} - \delta - \gamma \tilde{\mu} + \frac{1}{2} \gamma \tilde{\sigma}^2
$$
$$
\mu_{\zeta} = \frac{f'(t)}{f(t)} - \delta - \gamma \tilde{\mu} + \frac{1}{2} \gamma \tilde{\sigma}^2 + \frac{1}{2} \gamma^2 \tilde{\sigma}^2
$$
$$
\mu_{\zeta} = \frac{f'(t)}{f(t)} - \delta - \gamma \tilde{\mu} + \frac{1}{2} \gamma \tilde{\sigma}^2 (1+\gamma)
$$
从 $\frac{d\zeta_t}{\zeta_t} = -r_t \, dt - \lambda_t \, dz_t$ 我们知道 $\mu_{\zeta} = -r_t$ 和 $\sigma_{\zeta} = - \lambda_t$。

**短期利率 $r_t$：**
$$
r_t = - \mu_{\zeta} = \delta + \gamma \tilde{\mu} - \frac{1}{2} \gamma \tilde{\sigma}^2 (1+\gamma) - \frac{f'(t)}{f(t)}
$$
**市场风险价格 $\lambda_t$：**
$$
\lambda_t = - \sigma_{\zeta} = - (-\gamma \tilde{\sigma}) = \gamma \tilde{\sigma}
$$

**比较：**

没有习惯的“简单模型”、CRRA 效用 $u(c) = \frac{c^{1-\gamma}}{1-\gamma}$ 和消费 $c_t$ 遵循 GBM $\frac{dc_t}{c_t} = \mu \, dt + \sigma \, dz_t$。
在这种情况下，状态价格折现因子 $\zeta_t$ 为：
$$
\zeta_t = e^{-\delta t} u'(c_t) \cdot \text{常数} \propto e^{-\delta t} c_t^{-\gamma}
$$
使用同样的伊藤引理，我们可以得到：
$$
\frac{d\zeta_t}{\zeta_t} = \left( -\delta - \gamma \mu + \frac{1}{2} \gamma (1+\gamma) \sigma^2 \right) dt - \gamma \sigma \, dz_t
$$
在这种情况下，$\zeta_t$ 的漂移和扩散系数是**常数**。
* **短期利率 $r$** (常数):
    $$
    r = \delta + \gamma \mu - \frac{1}{2} \gamma (1+\gamma) \sigma^2
    $$
* **市场风险价格 $\lambda$** (常数):
    $$
    \lambda = \gamma \sigma
    $$

**比较结果：**

1.  **市场风险价格 $\lambda_t$：**
    * **习惯模型：** $\lambda_t = \gamma \tilde{\sigma}$。它是**常数**。
    * **简单模型：** $\lambda = \gamma \sigma$。它是**常数**。
    * **结论：** 形式上相同，但习惯模型中的 $\tilde{\sigma}$ 是**超额消费率 $\hat{c}_t$ 的波动率**，而简单模型中的 $\sigma$ 是**总消费 $c_t$ 的波动率**。如果 $\hat{c}_t \approx c_t$，则它们数值相近。

2.  **短期利率 $r_t$：**
    * **习惯模型：** $r_t = \delta + \gamma \tilde{\mu} - \frac{1}{2} \gamma \tilde{\sigma}^2 (1+\gamma) - \frac{f'(t)}{f(t)}$。它是**时变的**（依赖于 $t$）。
    * **简单模型：** $r = \delta + \gamma \mu - \frac{1}{2} \gamma (1+\gamma) \sigma^2$。它是**常数**。
    * **差异：**
        * 习惯模型中多了一个项 **$-\frac{f'(t)}{f(t)}$**。这一项反映了习惯形成带来的**时间期限效应**。
        * 由于 $f(t)$ 的形式包含 $e^{-\theta(T-t)}$，利率 $r_t$ 随时间 $t$ 变化（除非 $T \to \infty$）。
        * **如果 $T \to \infty$**，则 $e^{-\theta(T-t)} \to 0$。
            * $f(t) \to k (1 - \frac{\beta}{\theta})$，这是一个常数。
            * $f'(t) \to 0$。
            * 在这种情况下，**习惯模型中的 $r$ 也变为常数**： $r^{\infty} = \delta + \gamma \tilde{\mu} - \frac{1}{2} \gamma \tilde{\sigma}^2 (1+\gamma)$。
            * 即使在无限期情况下，习惯模型中的常数利率也与简单模型中的常数利率形式相似，但用**超额消费率的漂移率 $\tilde{\mu}$** 代替了**总消费的漂移率 $\mu$**，用**超额消费率的波动率 $\tilde{\sigma}$** 代替了**总消费的波动率 $\sigma$**。

在有限期限 $T$ 内，**习惯模型的短期利率是时变的**，这是习惯形成导致边际效用 (或其跨期扩展) 随时间变化的结果。
