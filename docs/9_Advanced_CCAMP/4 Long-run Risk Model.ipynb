# 长久期风险模型 (Long-run Risk Model)

长久期风险模型（Long-run Risk model，简称 LRR 模型）由 Bansal 和 Yaron 于 2004 年提出，它属于高级**基于消费的资产定价模型 (CCAPM)**，旨在通过修正**消费动态 (consumption dynamics)**来解决简单 CCAPM 无法匹配资产定价事实（例如股权溢价之谜和低无风险利率）的问题 。

### 核心思想和机制

LRR 模型的核心思想是，除了短期的、不可预测的消费冲击外，**预期的消费增长率**本身也存在一个**持久的 (persistent) 且时变的 (time-varying) 部分**，这个部分被称为“长久期风险” 。

该模型结合了**长久期消费风险**和 **Epstein-Zin 递归效用 (Epstein-Zin recursive utility)**，从而将**风险厌恶**和**跨期替代弹性 (EIS)**这两个偏好参数分离开来 。

在 LRR 模型中：

1.  **长期风险暴露：** 由于消费者具有 **Epstein-Zin 偏好**，他们对未来消费机会的担忧会转化为对**长期经济增长**的担忧。当预期的消费增长率（即长期风险因子 $x_t$）下降时，投资者的财富和未来消费机会会一同恶化，因此他们会要求更高的风险溢价来持有与此长期风险相关的资产 。
2.  **偏好参数要求：** 为了使模型成功解释股权溢价和低无风险利率，模型通常要求：
    * **相对风险厌恶系数 ($\gamma$ 或 $\vartheta$) 较高**：这使得投资者对风险非常敏感，从而推高风险溢价 。
    * **跨期替代弹性 (EIS，$\psi$ 或 $1/\alpha$) 较高（通常大于 1）**：这意味着投资者不喜欢消费波动，并愿意牺牲当前消费来平滑未来的消费流。当预期增长率 $x_t$ 下降时，未来的消费机会变差，投资者会增加储蓄（或减少借款），这会压低无风险利率 。
3.  **长期因子的持久性：** 模型中的长期风险因子 $x_t$ 被设定为具有**高持久性**（例如，月度估计的 $\rho$ 接近 1），这意味着预期的消费增长一旦变化，其影响将持续很长时间（例如，半衰期约为 33 个月） 。

### 模型的动态方程

LRR 模型中的**消费增长率**（对数形式 $g_{t+1} = \ln(C_{t+1}/C_t)$）和**长期因子**（$x_t$，预期的消费增长率）的动态方程通常设定为（Bansal & Yaron, 2004）:

* **消费增长率动态：**
    $$g_{t+1} = \mu + x_t + \sigma_t \eta_{t+1}$$
* **长期因子动态：**
    $$x_{t+1} = \rho x_t + \varphi_e \sigma_t e_{t+1}$$
* **消费波动率动态 (可选，随机波动率版本)：**
    $$\sigma_{t+1}^2 = \sigma^2 + \nu (\sigma_t^2 - \sigma^2) + \sigma_v v_{t+1}$$

其中：$\eta_{t+1}, e_{t+1}, v_{t+1}$ 是相互独立的 $N(0, 1)$ 冲击 。

* $\mu$: 消费增长的平均值。
* $x_t$: **时变的预期消费增长率**（长期风险因子），具有高持久性（$\rho$ 接近 1）。
* $\sigma_t$: **时变的消费波动率**（随机波动率版本中包含）。
* 当 $x_t \to 0$ 且 $\sigma_t \to \sigma$ 时，模型退化为简单的 CCAPM 。

### 模型的影响和评估

LRR 模型在金融学领域具有重要影响：

* **解决资产定价难题：** 校准后的模型参数能够产生与历史数据相符的**股权溢价**（例如，估计值 4.01%-6.84%，历史值 6.33%）和**低无风险利率**（例如，估计值 0.93%-1.44%，历史值 0.86%）。
* **无风险利率的波动：** 模型的无风险利率 $r_f$ 具有时变性，其波动性主要由预期的消费增长 $x_t$ 和消费波动率 $\sigma_t$ 的波动产生 。
* **风险溢价分解：** 股权溢价被分解为短期风险、长期风险和波动率风险溢价 。
* **局限性：** 尽管 LRR 模型在解释宏观资产定价事实方面取得了成功，但也面临一些挑战，例如：
    * **难以观测和估计：** 消费预期增长率 $x_t$ 难以直接观测 。
    * **与其他数据不符：** 难以匹配实际债券的定价 。
    * **代理人的时间偏好：** 校准的参数组合意味着代理人有巨大的“时间溢价” (timing premium)，即他们愿意放弃很大一部分消费来立即得知消费路径 。
