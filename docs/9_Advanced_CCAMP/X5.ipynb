## Solution

我们有一个离散时间经济模型，其中基于消费的附带 CAPM 成立，其公式为：

$E_t[r_{i,t+1}] = r_f + \beta_{it}(E_t[r_{C,t+1}] - r_f)$

其中：
- $E_t[r_{i,t+1}]$ 是资产 i 在时期 t 到 t+1 的预期净回报率。
- $r_f$ 是恒定的净无风险回报率，为 1% 或 0.01。
- $E_t[r_{C,t+1}]$ 是消费模拟投资组合的预期净回报率。
- $\beta_{it}$ 是资产 i 相对于消费模拟投资组合的时期 t beta 值。

经济在每个时期都处于两种状态之一：“好”或“坏”，且每种状态的概率均为 0.5。

---

### (a) 根据两种状态下 $\beta_{At}$, $\beta_{Bt}$, 和 $E_t[r_{C,t+1}]$ 的平均值，估算投资组合 A 和 B 的平均周期性回报率是多少？

首先，我们计算每种状态下变量的平均值：
- $\text{平均 } E_t[r_{C,t+1}] = (7\% + 11\%) / 2 = 9\% = 0.09$
- $\text{平均 } \beta_{At} = (-1 + 2) / 2 = 0.5$
- $\text{平均 } \beta_{Bt} = (3 + (-1)) / 2 = 1$

现在，我们将这些平均值代入 CAPM 公式来估算平均周期性回报率：

- $\text{投资组合 A 的估算平均回报率} = r_f + \text{平均 } \beta_{At}(\text{平均 } E_t[r_{C,t+1}] - r_f)$
- $\text{投资组合 A} = 1\% + 0.5(9\% - 1\%) = 1\% + 0.5(8\%) = 1\% + 4\% = 5\%$

- $\text{投资组合 B 的估算平均回报率} = r_f + \text{平均 } \beta_{Bt}(\text{平均 } E_t[r_{C,t+1}] - r_f)$
- $\text{投资组合 B} = 1\% + 1(9\% - 1\%) = 1\% + 8\% = 9\%$

---

### (b) 投资组合 A 在每种状态下的预期回报率是多少？投资组合 A 在两种状态下的正确平均回报率是多少？这与您在 (a) 中确定的平均回报率不同吗？

首先，我们计算投资组合 A 在两种状态下的预期回报率：
- **好状态：**
  $E_t[r_{A,t+1}] = r_f + \beta_{At}(\text{好} \text{状态}) (E_t[r_{C,t+1}] (\text{好状态}) - r_f)$
  $E_t[r_{A,t+1}] = 1\% + (-1)(7\% - 1\%) = 1\% - 6\% = -5\%$

- **坏状态：**
  $E_t[r_{A,t+1}] = r_f + \beta_{At}(\text{坏状态}) (E_t[r_{C,t+1}] (\text{坏状态}) - r_f)$
  $E_t[r_{A,t+1}] = 1\% + 2(11\% - 1\%) = 1\% + 2(10\%) = 1\% + 20\% = 21\%$

接下来，我们计算投资组合 A 在两种状态下的正确平均回报率。由于每种状态的概率都是 0.5，我们取两种状态下预期回报率的加权平均值：
- $\text{正确平均回报率 A} = 0.5 \times (\text{好状态回报率}) + 0.5 \times (\text{坏状态回报率})$
- $\text{正确平均回报率 A} = 0.5 \times (-5\%) + 0.5 \times (21\%) = -2.5\% + 10.5\% = 8\%$

这与 (a) 中确定的平均回报率 **不同**。在 (a) 中，估算的回报率是 5%，而正确的回报率是 8%。这个差异是由于**beta值和市场回报率的联合分布**造成的。

---

### (c) 投资组合 B 在每种状态下的预期回报率是多少？投资组合 B 在两种状态下的正确平均回报率是多少？这与您在 (a) 中确定的平均回报率不同吗？

首先，我们计算投资组合 B 在两种状态下的预期回报率：
- **好状态：**
  $E_t[r_{B,t+1}] = r_f + \beta_{Bt}(\text{好状态}) (E_t[r_{C,t+1}] (\text{好状态}) - r_f)$
  $E_t[r_{B,t+1}] = 1\% + 3(7\% - 1\%) = 1\% + 3(6\%) = 1\% + 18\% = 19\%$

- **坏状态：**
  $E_t[r_{B,t+1}] = r_f + \beta_{Bt}(\text{坏状态}) (E_t[r_{C,t+1}] (\text{坏状态}) - r_f)$
  $E_t[r_{B,t+1}] = 1\% + (-1)(11\% - 1\%) = 1\% - 1(10\%) = 1\% - 10\% = -9\%$

接下来，我们计算投资组合 B 在两种状态下的正确平均回报率：
- $\text{正确平均回报率 B} = 0.5 \times (\text{好状态回报率}) + 0.5 \times (\text{坏状态回报率})$
- $\text{正确平均回报率 B} = 0.5 \times (19\%) + 0.5 \times (-9\%) = 9.5\% - 4.5\% = 5\%$

这与 (a) 中确定的平均回报率 **不同**。在 (a) 中，估算的回报率是 9%，而正确的回报率是 5%。与投资组合 A 类似，这种差异也是由**beta值和市场回报率的联合分布**造成的。

---

### (d) 简要解释上述分析对消费型 CAPM（或任何其他附带资产定价模型）测试的启示。

上述分析的主要启示在于，在附带资产定价模型中，使用**无条件平均值**（即不考虑经济状态的平均值）来检验模型是**不正确的**，并且可能导致**错误的结论**。

具体来说，当资产的 beta 值随经济状态（例如，“好”状态或“坏”状态）变化时，且这些 beta 值与市场风险溢价（$E_t[r_C,t+1] - r_f$）**相关联**时，问题就会出现。

- 在我们的例子中，**投资组合 A** 的 beta 在坏状态下为正（2），而在好状态下为负（-1）。在坏状态下，市场风险溢价更高（10%），这使得 beta 为正时回报率更高。这种负相关性（即 beta 值在好的时候低，在坏的时候高）导致了错误的无条件平均值。

- 同样，**投资组合 B** 的 beta 在好状态下为正（3），而在坏状态下为负（-1）。当市场风险溢价在好状态下较低时（6%），这种正相关性（beta值在好的时候高，在坏的时候低）也导致了错误的无条件平均值。

因此，要**正确测试**消费型 CAPM 或任何其他附带资产定价模型，研究人员必须使用**附带测试**。这意味着他们需要**根据经济状态进行条件化**，并分别评估每种状态下的模型有效性。简单地用所有数据的平均值进行回归分析（类似于 (a) 中的计算）会忽略状态之间的差异，从而导致**模型似乎失效**，即使它在每种状态下都是有效的。这被称为**聚合偏差**或**状态相关性偏差**，是理解和测试这类模型时的一个关键挑战。
