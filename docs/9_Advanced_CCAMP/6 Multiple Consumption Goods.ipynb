# 多重消费品模型 (Multiple Consumption Goods)



在传统的基于**消费的资产定价模型 (CCAPM)** 中，通常假设存在一个**单一的、可加总的消费品**，投资者对这个总消费流的风险厌恶程度决定了资产价格。然而，**多重消费品模型**通过引入**多于一种消费品**，并允许投资者对不同消费品或消费品组别表现出**异质的偏好 (Heterogeneous Preferences)**，来扩展和改进 CCAPM。

这种扩展的主要目标是：

1.  **更准确地描述消费者的效用：** 消费篮子中的不同商品（例如，耐用品、非耐用品、必需品、奢侈品）可能以不同的方式影响效用，并且它们的通货膨胀率（价格）和消费量可能具有不同的动态。
2.  **解释横截面资产收益：** 尝试解释为什么某些类型的资产（例如，对必需品价格敏感的股票）与另一些资产（例如，对奢侈品消费敏感的股票）具有不同的风险溢价。

### 核心机制和类型

多重消费品模型主要有两种引入方式：

#### 1. 耐用品与非耐用品 (Durables vs. Non-durables)

这是最常见的类型，由 Dunn 和 Singleton (1986)、Eichenbaum 和 Hansen (1990) 等人提出。

* **问题所在：** 简单的 CCAPM 使用**非耐用品和劳务**的消费数据（例如，食品、服务），这些数据的波动性很低。但**耐用品 (Durables)** 的购买（如汽车、家电）更像是**投资决策**，波动性高得多。
* **机制：** 如果将耐用品视为一种**资产**，其**服务流**（即耐用品带来的效用）才是真正的消费。
    * 模型将消费效用定义为**非耐用品消费**和**耐用品存量**的服务流的函数。
    * 由于耐用品购买具有高度周期性，且在经济衰退时大幅下降，它提供了更具**波动性和前瞻性**的消费度量，能够产生更高的**边际效用波动 (Volatility of Marginal Utility)**。
* **结果：** 引入耐用品可以**显著提高**边际效用的波动性，从而在一定程度上**提高模型预测的股权溢价**，缓解**股权溢价之谜**。

#### 2. 必需品与奢侈品/可支配消费 (Necessities vs. Luxuries/Discretionary Consumption)

这类模型关注不同消费品对投资者**效用函数**的差异化影响，通常结合**习惯形成 (Habit Formation)** 或**状态依赖偏好**。

* **模型假设：** 假设消费品分为两类：
    * **必需品 (Necessities)：** 消费相对稳定，其消费量对经济状况不敏感。
    * **奢侈品/可支配消费 (Luxuries)：** 消费量对经济环境（特别是对**财富**和**收入**）高度敏感，在经济衰退时下降更多。
* **状态依赖偏好 (State-Dependent Preferences)：** 投资者可能只对**可支配消费**的风险表现出高风险厌恶。例如，当经济状况良好时，他们更关注奢侈品消费的乐趣；当经济衰退时，他们会大幅削减奢侈品，这使得**边际替代率 (MRS)** 对奢侈品消费的变化更加敏感。
* **结果：**
    * **高风险溢价：** 资产收益与**奢侈品消费增长**高度正相关。当经济衰退时，奢侈品消费暴跌，同时股票收益也暴跌，使得股票是**坏状态下的坏资产**，因此要求高溢价。
    * **横截面定价：** 这个框架为**三因子模型**中的**市值效应 (Size Effect)** 提供了一种可能的解释：小盘股的收入对可支配消费的敏感性可能更高，因此其风险溢价更高。

### 总结

多重消费品模型通过细化消费篮子，使得消费动态的设定更加灵活和真实，从而能够更好地与资产价格的事实相匹配：

| 模型类型 | 关注焦点 | 主要机制 | 对资产定价的影响 |
| :--- | :--- | :--- | :--- |
| **耐用品模型** | 消费的波动性 | 耐用品购买具有高波动性，是消费的更具前瞻性的度量。 | **提高**边际效用波动，有助于解决**股权溢价之谜**。 |
| **必需品/奢侈品模型** | 消费的敏感性 | 效用对**奢侈品/可支配消费**的变动更敏感。 | 资产收益与奢侈品消费的共同变化提供了**系统性风险**来源，可解释横截面收益差异。 |
