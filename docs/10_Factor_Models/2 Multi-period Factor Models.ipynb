# 多期因子模型 (Multi-period Factor Models)



多期因子模型主要关注**条件 (Conditional)** 和**非条件 (Unconditional)** 模型的区别，以及如何从理论模型（如 Consumption-CAPM 和 Intertemporal CAPM）中推导出多期因子关系 。

---

## 离散时间：条件 vs. 非条件模型

在离散时间框架下，因子模型可以分为条件模型和非条件模型 ：

### 1. 条件因子模型 (Conditional Factor Model)

条件因子模型是**给定时间 $t$ 的信息**来解释下一期 $t+1$ 的预期收益 。

* **关系式**（简化形式）：
    $$E_{t}[R_{i,t+1}] = \alpha_{t} + \beta_{t}[R_{i,t+1}, x_{t+1}]^{\top}\eta_{t}$$
    其中 $\beta_{t}$ 表示**条件 $\beta$**，$\eta_{t}$ 表示**条件风险溢价**，它们是**随时间变化的** 。
* 条件因子模型几乎等价于**状态价格折现因子 (SPD)** 的比率 $\zeta_{t+1}/\zeta_{t}$ 是**因子 $x_{t+1}$ 的仿射或指数仿射函数** 。

### 2. 非条件因子模型 (Unconditional Factor Model)

非条件因子模型是**对所有时间**来解释**平均预期收益** 。

* **关系式**：
    $$E[R_{i,t+1}] = \alpha + \beta[R_{i,t+1}, x_{t+1}]^{\top}\eta$$
* 这种模型对于**检验模型或使用历史数据估计模型参数**（如 $\beta$ 值）是必要的 。

### 从条件到非条件模型的转换

* **问题**：如果条件因子模型成立，非条件模型**不一定成立** 。
* **推导**：通过取条件模型 $E_{t}[R_{i,t+1}] = R_{f,t} + \beta_{i,t}\vartheta_{t}$ 的非条件期望，可以发现：
    $$E[R_{i,t+1}]= E[R_{f,t}]+E[\beta_{i,t}]E[\vartheta_{t}] + Cov [\beta_{i,t} , \vartheta_{t}]$$
    **协方差项 $Cov [\beta_{i,t} , \vartheta_{t}]$ 的存在**是导致条件 CAPM 成立而简单的非条件 CAPM 不成立的关键 。
* **扩大非条件因子**：如果 $\zeta_{t+1}/\zeta_{t} = a_{t} + b_{t}^{\top}x_{t+1}$，且系数 $a_{t}, b_{t}$ 本身是**工具变量 (instrument vector)** $y_{t}$ 的函数（如 $a_{t} = A_{0} + A_{1}y_{t}$），那么**新的非条件因子模型**将包含原始因子 $x_{i}$、工具变量 $y_{j}$ **以及它们的交互项** $x_{i}y_{j}$ 。工具向量 $y_{t}$ 应该反映单期 SPD 中**条件矩的变化** 。

---

## 连续时间因子模型

在连续时间框架下，不需要假设 SPD 是因子的仿射或指数仿射函数 。

* 资产的**风险溢价**（超额瞬时预期收益 $\mu_{i,t} + \frac{1}{2}\sigma_{i,t}^{2} - r_{f,t}$）可以表示为**因子暴露**和**因子风险溢价**的乘积形式 。
* 这也会导出一个**条件多因子模型** 。

## 应用：CCAPM 和 Intertemporal CAPM (ICAPM)

### 1. Consumption-CAPM (CCAPM)

CCAPM 天然是一个**条件模型** 。

* **核心因子**：**消费增长率** $g_{t+1} = (c_{t+1} - c_{t})/c_{t}$ 。
* **问题**：简单的/非条件的 CCAPM **表现不佳**，它无法解释股票收益的**水平**（股权溢价之谜）或解释不同股票**平均收益的差异** 。
* **改进**：**条件 CCAPM** 在解释收益的横截面差异方面表现更好，例如：
    * Lettau & Ludvigson (2001) 使用**对数消费/财富比率 ($cay_{t}$)** 作为工具变量，因为它反映了投资者对未来投资机会的预期 。

### 2. Intertemporal CAPM (ICAPM)

ICAPM 是对 CAPM 的多期推广，它假设投资者不仅关心下一期的**财富收益**，还关心**投资机会集的变化**。

* **核心因子**：ICAPM 导出一个**近似的条件双因子模型**：
    $$E_{t}[R_{i,t+1}] \approx \alpha_{t} + \beta_{i,W,t}\vartheta_{W,t} + \beta_{i,X,t}\vartheta_{X,t}$$
    其中，因子通常是：
    1.  **市场组合（财富 $W$）的收益**（或财富变化）。
    2.  **额外的状态变量 $X$**，它们捕捉**投资机会的变化** 。
* **结论**：**只有**当投资机会没有变化，**或者**投资者不关心（如对数效用），ICAPM 才会退化为一个**条件单因子 CAPM**。ICAPM **不支持**一个非条件的**单因子** CAPM 。
* **实证发现**：
    * Jagannathan & Wang (1996) 证明条件 CAPM 可以转化为一个包含**市场 $\beta$** 和**$\beta$-溢价敏感性**的**非条件双因子模型**，这比传统的非条件 CAPM 解释力更强。他们进一步加入**人均劳动收入增长率**作为第三个因子，以代理**人力资本**，解释力进一步提高 。
    * Brennan, Wang & Xia (2004) 提出了一个包含**短期利率**和**资本市场线斜率**的 **3 因子 ICAPM** 。
