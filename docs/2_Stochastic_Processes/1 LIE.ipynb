# Law of Iterated Expectation

> Thereom(Law of Iterated Expectations)
for $t<t'$,
> $$ E_t[Et'[X]] = E_t[X]$$


期望的迭代定律是概率论中的一个基本定理，它表明对一个随机变量进行分步（或多层）条件期望的计算，结果与一步到位的总期望是相等的。

对于随机变量 $X$ 以及信息集合 $\mathcal{F}_t$（$t$ 时刻已知的信息）和 $\mathcal{F}_{t+1}$（$t+1$ 时刻已知的信息），其中 $\mathcal{F}_t \subset \mathcal{F}_{t+1}$（$t+1$ 知道的比 $t$ 多），LIE 可以表述为：

$$\mathbf{E}_t [ X ] = \mathbf{E}_t [ \mathbf{E}_{t+1} [ X ] ]$$

* $\mathbf{E}_t [ X ]$: 在 $t$ 时刻已知的信息下，对 $X$ 的期望。
* $\mathbf{E}_{t+1} [ X ]$: 在 $t+1$ 时刻已知的信息下，对 $X$ 的期望。**注意：** 这是一个**随机变量**，因为 $t+1$ 时刻的状态是不确定的。
* $\mathbf{E}_t [ \mathbf{E}_{t+1} [ X ] ]$: 对 $t+1$ 时刻的这个**随机变量**再在 $t$ 时刻求期望。

**通俗地讲：** “总期望等于分步期望的期望”。
