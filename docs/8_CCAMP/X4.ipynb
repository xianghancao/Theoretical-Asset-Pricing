## Idea

- formula of expectation & variance
- next-period SPD
- discrete-time $r^f_t$ 满足 $r^f = 1/E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$
-  continuously compounded risk-free rate $e^{r^f} = 1/E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$

## Solution


### a) 计算 $E_t\left[ \frac{C_{t+1}}{C_t} \right]$ 和 $Var_t\left[ \frac{C_{t+1}}{C_t} \right]$

根据题意，我们知道：
- 当概率为1/2时，$\frac{C_{t+1}}{C_t} = (1+\alpha)\mu$
- 当概率为1/2时，$\frac{C_{t+1}}{C_t} = (1-\alpha)\mu$

首先计算期望值 $E_t\left[ \frac{C_{t+1}}{C_t} \right]$：
$E_t\left[ \frac{C_{t+1}}{C_t} \right] = \frac{1}{2} \cdot (1+\alpha)\mu + \frac{1}{2} \cdot (1-\alpha)\mu = \frac{1}{2}\mu(1+\alpha+1-\alpha) = \mu$

接着计算方差 $Var_t\left[ \frac{C_{t+1}}{C_t} \right]$：
$Var_t\left[ \frac{C_{t+1}}{C_t} \right] = E_t\left[ \left(\frac{C_{t+1}}{C_t}\right)^2 \right] - \left(E_t\left[ \frac{C_{t+1}}{C_t} \right]\right)^2$

先计算 $E_t\left[ \left(\frac{C_{t+1}}{C_t}\right)^2 \right]$：
$E_t\left[ \left(\frac{C_{t+1}}{C_t}\right)^2 \right] = \frac{1}{2} \cdot \left((1+\alpha)\mu\right)^2 + \frac{1}{2} \cdot \left((1-\alpha)\mu\right)^2$
$= \frac{1}{2}\mu^2 \left( (1+2\alpha+\alpha^2) + (1-2\alpha+\alpha^2) \right) = \frac{1}{2}\mu^2(2+2\alpha^2) = \mu^2(1+\alpha^2)$

所以，方差 $Var_t\left[ \frac{C_{t+1}}{C_t} \right]$ 为：
$Var_t\left[ \frac{C_{t+1}}{C_t} \right] = \mu^2(1+\alpha^2) - \mu^2 = \mu^2\alpha^2$

**结果解释**：
- 期望值 $E_t\left[ \frac{C_{t+1}}{C_t} \right] = \mu$ 表示下一期消费增长率的平均值是 $\mu$。
- 方差 $Var_t\left[ \frac{C_{t+1}}{C_t} \right] = \mu^2\alpha^2$ 表示消费增长率围绕其均值 $\mu$ 的波动程度，波动的大小由 $\mu$ 和 $\alpha$ 决定。其中，$\alpha$ 越大，波动性越强。

---

### b) 状态价格平减指数和 $E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$

在代表性消费者模型中，边际替代率（Marginal Rate of Substitution, MRS）是状态价格平减指数 $\frac{\zeta_{t+1}}{\zeta_t}$ 的倒数。对于幂效用函数 $U(C) = \frac{C^{1-\gamma}}{1-\gamma}$，边际效用为 $U'(C) = C^{-\gamma}$。
状态价格平减指数可以表示为：
$\frac{\zeta_{t+1}}{\zeta_t} = \beta \frac{U'(C_{t+1})}{U'(C_t)} = e^{-\delta} \frac{C_{t+1}^{-\gamma}}{C_t^{-\gamma}} = e^{-\delta} \left(\frac{C_{t+1}}{C_t}\right)^{-\gamma}$
其中 $\beta = e^{-\delta}$。

根据 a) 的结果，$\frac{C_{t+1}}{C_t}$ 有两种可能取值，因此 $\frac{\zeta_{t+1}}{\zeta_t}$ 也有两种可能取值：
- 当概率为1/2时，$\frac{\zeta_{t+1}}{\zeta_t} = e^{-\delta} ((1+\alpha)\mu)^{-\gamma}$
- 当概率为1/2时，$\frac{\zeta_{t+1}}{\zeta_t} = e^{-\delta} ((1-\alpha)\mu)^{-\gamma}$

接下来计算 $E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$：
$E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = \frac{1}{2} e^{-\delta} ((1+\alpha)\mu)^{-\gamma} + \frac{1}{2} e^{-\delta} ((1-\alpha)\mu)^{-\gamma}$
$= \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)$

---

### c) 证明连续复利无风险利率的表达式

无风险利率 $r^f$ 满足等式 $e^{-r^f} = E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$。
利用 b) 的结果，我们有：
$e^{-r^f} = \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)$

对等式两边取自然对数：
$-r^f = \ln\left[ \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right) \right]$
$-r^f = \ln\left( \frac{1}{2} \right) + \ln\left( e^{-\delta} \right) + \ln\left( \mu^{-\gamma} \right) + \ln\left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)$
$-r^f = -\ln(2) - \delta - \gamma\ln(\mu) + \ln\left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)$

将 $-r^f$ 移项，并整理得到：
$r^f = \delta + \gamma\ln(\mu) - \left[ \ln\left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right) - \ln(2) \right]$
$r^f = \delta + \gamma\ln(\mu) - \ln\left( \frac{(1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma}}{2} \right)$

这与题目中给出的表达式完全一致。

---

### d) 证明最大夏普比率的表达式

最大夏普比率的公式为 $SR_{max} = \frac{\sigma_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]}{E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]}$。
这里 $\sigma_t$ 表示条件标准差，即 $\sigma_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = \sqrt{Var_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]}$。
首先计算 $Var_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$：
$Var_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = E_t\left[ \left(\frac{\zeta_{t+1}}{\zeta_t}\right)^2 \right] - \left(E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]\right)^2$

根据 b) 中 $\frac{\zeta_{t+1}}{\zeta_t}$ 的可能值，计算 $E_t\left[ \left(\frac{\zeta_{t+1}}{\zeta_t}\right)^2 \right]$：
$E_t\left[ \left(\frac{\zeta_{t+1}}{\zeta_t}\right)^2 \right] = \frac{1}{2} \left(e^{-\delta} ((1+\alpha)\mu)^{-\gamma}\right)^2 + \frac{1}{2} \left(e^{-\delta} ((1-\alpha)\mu)^{-\gamma}\right)^2$
$= \frac{1}{2} e^{-2\delta} \mu^{-2\gamma} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} \right)$

将 b) 中的 $E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]$ 代入方差公式：
$Var_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = \frac{1}{2} e^{-2\delta} \mu^{-2\gamma} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} \right) - \left[ \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right) \right]^2$
$= e^{-2\delta} \mu^{-2\gamma} \left[ \frac{1}{2} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} \right) - \frac{1}{4} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)^2 \right]$
$= e^{-2\delta} \mu^{-2\gamma} \left[ \frac{1}{2} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} \right) - \frac{1}{4} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} + 2(1+\alpha)^{-\gamma}(1-\alpha)^{-\gamma} \right) \right]$
$= e^{-2\delta} \mu^{-2\gamma} \left[ \frac{1}{4} \left( (1+\alpha)^{-2\gamma} + (1-\alpha)^{-2\gamma} - 2(1-\alpha^2)^{-\gamma} \right) \right]$
$= \frac{1}{4} e^{-2\delta} \mu^{-2\gamma} \left[ (1+\alpha)^{-\gamma} - (1-\alpha)^{-\gamma} \right]^2$

所以，$\sigma_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = \sqrt{Var_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]} = \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left| (1+\alpha)^{-\gamma} - (1-\alpha)^{-\gamma} \right|$。
因为 $\alpha>0, \gamma>0$ 且 $(1-\alpha)^{-\gamma} > (1+\alpha)^{-\gamma}$，所以绝对值可以去掉。
$\sigma_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right] = \frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1-\alpha)^{-\gamma} - (1+\alpha)^{-\gamma} \right)$

最后计算 $SR_{max}$：
$SR_{max} = \frac{\sigma_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]}{E_t\left[ \frac{\zeta_{t+1}}{\zeta_t} \right]} = \frac{\frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1-\alpha)^{-\gamma} - (1+\alpha)^{-\gamma} \right)}{\frac{1}{2} e^{-\delta} \mu^{-\gamma} \left( (1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma} \right)}$
$SR_{max} = \frac{(1-\alpha)^{-\gamma} - (1+\alpha)^{-\gamma}}{(1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma}}$

这与题目中给出的表达式一致。

---

### e) 利用给定数值计算 $r^f$ 和 $SR_{max}$

给定数值为：$\mu=1.02$, $\alpha=0.06$, $\delta=0.01$, $\gamma=10$。
首先计算连续复利无风险利率 $r^f$。使用 c) 中的公式：
$r^f = \delta + \gamma\ln(\mu) - \ln\left( \frac{(1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma}}{2} \right)$
代入数值：
$r^f = 0.01 + 10\ln(1.02) - \ln\left( \frac{(1+0.06)^{-10} + (1-0.06)^{-10}}{2} \right)$
$r^f = 0.01 + 10\ln(1.02) - \ln\left( \frac{(1.06)^{-10} + (0.94)^{-10}}{2} \right)$
使用计算器计算：
$\ln(1.02) \approx 0.01980$
$(1.06)^{-10} \approx 0.55839$
$(0.94)^{-10} \approx 1.83921$
$r^f \approx 0.01 + 10(0.01980) - \ln\left( \frac{0.55839 + 1.83921}{2} \right)$
$r^f \approx 0.01 + 0.1980 - \ln(1.1988) \approx 0.2080 - 0.1812 \approx 0.0268$
所以，连续复利无风险利率约为 **0.0268**。

接着计算最大夏普比率 $SR_{max}$。使用 d) 中的公式：
$SR_{max} = \frac{(1-\alpha)^{-\gamma} - (1+\alpha)^{-\gamma}}{(1+\alpha)^{-\gamma} + (1-\alpha)^{-\gamma}}$
代入数值：
$SR_{max} = \frac{(0.94)^{-10} - (1.06)^{-10}}{(1.06)^{-10} + (0.94)^{-10}}$
$SR_{max} = \frac{1.83921 - 0.55839}{0.55839 + 1.83921} = \frac{1.28082}{2.3976} \approx 0.5342$
所以，最大夏普比率约为 **0.5342**。
