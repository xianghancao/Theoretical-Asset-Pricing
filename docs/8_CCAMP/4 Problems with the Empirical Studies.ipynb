# 经验研究的挑战 (Problems with the Empirical Studies)

对 CCAPM 进行经验检验时，研究人员面临着几个与数据质量和统计方法相关的重大挑战 ：

## 1. 总消费数据的质量问题 (Aggregate Consumption Data Quality)

消费数据是 CCAPM 的核心，但其质量存在严重缺陷 ：

* **时间汇总 (Temporally Aggregated)**：国家收入与产品账户 (NIPA) 的总消费数据是按季度报告的，这意味着它汇总了三个月的数据。这种**时间汇总**使得估计实际的边际效用（这是一个瞬时概念）变得困难 。
* **衡量购买而非消费 (Purchases vs. Consumption)**：数据显示的是**购买**（尤其是耐用品和可储存商品），而不是实际的**消耗** 。例如，购买一辆车被计入消费，但它的效用是在多年内消耗的。
* **测量误差 (Measurement Errors)**：消费数据可能包含较大的测量误差，并且经常在事后被修订 。
* **不精确和不频繁采样 (Infrequent Sampling)**：消费数据不如资产价格数据那样频繁和精确 。

由于这些问题，一些研究人员转向使用**调查消费数据**，这些数据显示出更高的消费波动性，这有助于**缩小**股权溢价之谜 。此外，也有研究者探索使用更具创造性的消费衡量指标，例如 **垃圾量** (garbage as a measure of consumption)，据说可以更好地匹配股票回报和无风险利率 。

## 2. 平均超额股票回报的估计不精确 (Imprecise Estimation of Mean Excess Stock Return)

* **统计误差**：股票的平均超额回报（风险溢价）是一个难以精确估计的量 。
* **宽泛的置信区间**：根据历史数据（例如 50 个观测值），年度股票超额回报的标准误差很高（如 $\sigma / \sqrt{T} \approx 2.8\%$），导致 $95\%$ 置信区间非常宽（例如 $[2.5\%, 13.5\%]$）。
* **影响**：如果理论上的股权溢价是 $2.5\%$ 而非经验观察到的 $8\%$，那么**股权溢价之谜在量级上会小得多** 。

## 3. 幸存者偏差和 Peso 问题 (Survivorship Bias and Peso Problem)

* **幸存者偏差**：用于计算股权溢价的历史数据往往来自成功的市场（如美国）。在 1950 年，投资者并不知道美国经济会避免重大萧条或战争，这导致了**实现的**回报高于**预期的**回报 。
* **Peso 问题**：部分风险溢价可能是由于**未实现**的低概率灾难风险所致（即“Peso 问题”）。投资者预期到低概率、高影响的事件（如金融危机或大萧条），但这些事件并未在样本期内发生。
* **影响**：稀有灾难模型表明，这种未实现的风险可能每年额外解释 $2\%$ 到 $4\%$ 的回报率 。

## 4. 缺乏数据平稳性 (Lack of Data Stationarity)

* **理论要求**：使用单一的已实现回报和消费序列来检验关于事前预期 (ex-ante expectations) 的理论，要求数据是**平稳的**，即所有观测值都来自相同的概率分布 。
* **现实挑战**：由于以下因素，数据可能不是平稳的 ：
    * **违约风险**：如果资产发生违约（回报为 $-100\%$），未来的回报甚至可能变得不确定 。
    * **政策变化**：在过去 50 年中，美国税收规则的重大变化可能有利于股票投资，从而影响所需风险溢价 。
    * **交易成本**：交易成本和约束的减少也可能降低所需的风险溢价 。

## 5. 预期形成问题 (Forming Expectations)

* **主观预期**：理论模型涉及个体的主观预期，但投资者的预期可能系统性地过于乐观或过于悲观 。
* **行为金融学**：如果投资者系统性地过于悲观，资产价格就会低估，导致实现的年回报率偏高 。

---

正是这些实证上的挑战和模型的局限性，导致了 CCAPM 的进一步发展，例如转向**条件 CCAPM**（允许参数随时间变化）和开发**因子模型**（如 Fama-French 三因子模型）作为解释股票回报的替代方法 。
