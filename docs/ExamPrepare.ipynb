**考试信息:**
* **课程:** Theoretical Asset Pricing (理论资产定价)
* **时长:** 4小时
* **题目数量:** 4-5个

**材料分析:**

1.  **FormulaSheet (公式表):** 涵盖范围非常广，从基础的随机过程 (布朗运动, Itô 过程/引理, GBM) 到核心定价理论 (无套利, LOP, 状态价格, SPD, 风险中性定价)，再到效用理论、个体最优、均衡理论 (帕累托最优, 代表性个人, 福利经济学第一定理)，然后是具体的模型应用，如 CCAPM (离散/连续时间)，因子模型，最后是利率期限结构模型 (Vasicek)。还包括了一些概率论、微积分和线性代数的基础知识。这表明课程内容全面且深入。
2.  **Problem Set 1 Solutions:**
    * **X-1:** 应用 Itô 引理 (二维)，计算瞬时均值、方差、协方差、相关性，识别过程类型 (GBM, ABM)。**核心技能：随机微积分应用。**
    * **3.4:** 离散状态定价，求解状态价格，检查无套利，构造套利组合。**核心技能：状态定价基础，套利原理。**
    * **X-2:** 离散状态+幂效用，计算 SPD，无风险利率，资产定价，风险溢价，分析灾难状态影响。**核心技能：基于效用的定价，计算关键指标。**
    * **4.6:** SPD 理论，错误定价模型分析，利用柯西-施瓦茨不等式推导错误定价边界。**核心技能：SPD理论，协方差性质。**
    * **4.10:** 多期离散定价，计算条件期望($E_t[\frac{D_{t+1}}{D_t}\frac{\zeta_{t+1}}{\zeta_t}]$)，应用迭代期望定律，几何级数求和得到价格，参数敏感性分析。**核心技能：多期定价，迭代期望定律，对数正态性质。**
    * **4.11:** 连续时间零息债券定价 ($B_t^s = E_t[\zeta_s/\zeta_t]$)，假设 SPD 为 GBM (等价于常数 $r^f, \lambda$)，计算收益率和收益率曲线。**核心技能：连续时间债券定价。**
    * **X-3:** 两期树状模型，计算状态依赖的无风险利率，动态规划定价，解释价格差异 (风险相关性)，检查市场完备性。**核心技能：树状模型，动态规划定价，市场完备性。**
    * **5.5:** 效用理论细节，风险厌恶与凹性，绝对/相对风险厌恶 (ARA/RRA) 与对赌博规模敏感度的关系 (涉及泰勒展开/极限)。**核心技能：效用理论，风险厌恶度量。**
    * **6.3:** 不同效用函数下的最优选择和状态定价 ($\psi_\omega \propto \text{MRS}$)。**核心技能：效用函数应用。**
3.  **Problem Set 2 Solutions:**
    * **7.1:** 证明加权平均 SPD 仍是 SPD，完备市场与代表性个人的关系。**核心技能：SPD性质，均衡与代表性个人。**
    * **8.4:** 连续时间 CCAPM，但消费过程非标准 GBM。推导风险市场价格 $\lambda_t$，超额收益，无风险利率 $r^f_t$，分析参数影响 (周期性)。应用 Itô 引理推导 $dr^f_t$。**核心技能：CCAPM 应用，Itô 引理，模型分析。**
    * **8.5:** 离散时间 CCAPM，SPD 与消费增长呈线性关系。校准参数，多期债券定价，动态规划定价含股息资产，解释价格差异。**核心技能：离散 CCAPM，参数校准，动态定价。**
    * **X-4:** 离散时间 CCAPM，二项式消费增长。计算 SPD 矩，推导无风险利率，推导最大夏普比率。**核心技能：离散 CCAPM，矩计算，夏普比率。**
    * **X-5:** 条件 CCAPM vs 非条件 CCAPM，说明忽略时变性对实证检验的误导。**核心技能：条件模型 vs 非条件模型。**
    * **X-6:** 连续时间习惯形成模型 (内部习惯)。推导习惯动态，推导 SPD (含积分项)，在特定假设下简化 SPD，推导 $r^f_t$ 和 $\lambda_t$。**核心技能：习惯形成模型，求解积分，Itô 引理。**
    * **9.4:** 连续时间外部习惯模型 ("攀比")。定义剩余消费比，推导 SPD，应用 Itô 引理 (多变量) 推导 SPD 动态，识别 $r^f_t$ 和 $\lambda_t$，推导因子模型形式的超额收益，分析夏普比率周期性。**核心技能：外部习惯模型，多变量 Itô 引理，因子模型联系，模型分析。**
    * **10.3:** 离散时间线性因子模型。直接用 SPD 定价 vs 用 beta 定价公式定价。推导 beta 定价所需各部分 (期望，协方差，方差，beta，风险溢价因子)，验证等价性。**核心技能：因子模型，两种定价方法比较，矩计算。**
    * **11.8:** 连续时间 Vasicek 模型背景下，对未来现金流定价。证明期望是鞅，推导索赔价值 $V_t = E_t[\frac{\zeta_T}{\zeta_t} F_T]$，计算风险调整贴现率 $R_t$，与无风险收益率 $y_t^T$ 比较，分析风险溢价期限结构。用 Itô 引理验证 $V_t$ 满足无套利条件。**核心技能：Vasicek 应用，或有索赔定价，风险调整贴现率，Itô 引理验证。**

**核心主题总结:**

* **随机微积分:** Itô 引理是核心工具，必须熟练掌握单变量和多变量应用。
* **定价基础:** 无套利、状态价格、SPD、风险中性定价之间的关系，及其在离散和连续时间的应用。
* **CCAPM:** 基于消费的定价是重点，包括标准模型和扩展（如习惯形成）。能够推导关键变量（$r^f, \lambda_t$, 超额收益）并分析模型含义。
* **均衡概念:** 帕累托最优、代表性个人、市场完备性及其关系。
* **利率期限结构:** 特别是 Vasicek 这种仿射模型，债券定价和收益率曲线。
* **数学功底:** 微积分、概率论（特别是正态/对数正态分布的期望）、线性代数。

**可能的考试题型预测:**

基于以上分析，4-5个题目很可能会围绕以下几种类型展开，通常一个大题会包含多个小问，融合多个知识点：

1.  **随机过程与 Itô 引理应用题 (必考类型，可能1-2题):**
    * 给定几个基础随机过程的 SDE (例如股价、利率、波动率、消费等)。
    * 要求推导某个衍生变量（例如期权价格、债券价格、财富、效用函数值、两种资产的比率/乘积）的 SDE。这会直接考察 Itô 引理的应用，可能涉及多维或复杂函数。
    * 计算瞬时期望、方差、协方差、相关性。
    * 求解简单的 SDE (如 GBM, OU 过程)。
    * *类似习题:* PS1: X-1, PS2: 8.4(c), PS2: X-6(a,c,d), PS2: 9.4(c), PS2: 11.8(c)。

2.  **资产定价基础与均衡 (离散状态空间) (可能1题):**
    * 给定一个简单的多期（如两期）或单期多状态的经济设定（禀赋、资产支付 D 矩阵）。
    * 判断是否存在套利，计算状态价格 $\psi$ 或风险中性概率 $q$。
    * 判断市场是否完备 (rank(D) vs S)，讨论不完备市场的影响（状态价格/风险中性概率不唯一）。
    * 使用状态价格或风险中性概率为新的衍生品定价。
    * 可能结合简单的效用函数，计算均衡价格和分配。
    * *类似习题:* PS1: 3.4, PS1: X-3, PS1: 6.3。

3.  **CCAPM 推导与分析 (必考类型，可能1-2题):**
    * 给定代表性代理人的效用函数（很可能是幂效用，但也可能是习惯形成等扩展形式）和消费过程的动态（离散或连续 SDE）。
    * 推导状态价格偏转因子 (SPD) 的动态过程 $d\zeta_t / \zeta_t$ 或离散形式 $\zeta_{t+1}/\zeta_t$。
    * 推导均衡下的无风险利率 $r^f_t$ 和风险的市场价格 $\lambda_t$。解释各组成部分的经济含义（如 PS2: 8.4 的周期性分析）。
    * 推导一般资产的预期超额收益公式 $(\mu_i + \delta_i) - r^f = \text{Cov}(...) = \sigma_i^T \lambda_t$。
    * 计算特定资产（如股权，代表消费索取权）的风险溢价或夏普比率。
    * 讨论模型的含义，比如能否解释金融市场异象（权益溢价之谜等）。
    * *类似习题:* PS1: X-2, PS1: 4.10, PS2: 7.1, PS2: 8.4, PS2: 8.5, PS2: X-4, PS2: X-6, PS2: 9.4。

4.  **利率期限结构模型 (很可能1题):**
    * 给定一个短期利率模型（大概率是 Vasicek，因为公式表里有，习题也涉及）。
    * 写出或推导零息债券价格 $B(t,T)$ 的表达式（可能要求推导仿射形式下的 ODEs）。
    * 计算不同期限的收益率 $y_t^T$ 和瞬时远期利率 $f_t^T$。
    * 分析模型预测的收益率曲线形状及其动态。
    * 可能要求计算债券的预期收益和风险溢价。
    * 可能结合 CCAPM 解释利率动态的来源。
    * *类似习题:* PS2: 11.8 (虽然是定价问题，但背景是 Vasicek)。

5.  **理论辨析与证明题 (可能融入其他题目或单独出现):**
    * 阐述无套利、单一价格定律、状态价格存在性/唯一性/正性、SPD、风险中性测度、市场完备性之间的逻辑关系。
    * 证明或解释福利经济学第一定理。
    * 解释代表性代理人存在性的条件。
    * 讨论条件模型与非条件模型的区别及其在实证检验中的意义。
    * *类似习题:* PS1: 4.6, PS1: X-3(b,c), PS2: 7.1(b), PS2: X-5(d)。

**备考建议:**

* **掌握核心推导:** 务必熟练掌握 Itô 引理、SPD 的推导（基于效用）、CCAPM 定价方程的推导、Vasicek 模型债券定价的推导。这些是构建模型和解决问题的基础。
* **理解经济直觉:** 不要只记忆公式，要理解每个公式背后的经济含义，例如 $r^f$ 公式的三个组成部分，$\lambda_t = \gamma \sigma_c$ 的含义，夏普比率与 SPD 的关系等。这有助于解题和回答理论辨析题。
* **熟悉公式表:** 考试时公式表是你的重要工具，平时练习就要熟悉公式的位置和适用条件。
* **练习计算:** 习题中涉及大量的计算（期望、方差、协方差、条件期望、求解方程组、应用对数正态性质等），要保证计算的准确性和速度。
* **注意离散与连续的转换:** 理解核心概念在两种框架下的对应关系（如求和 vs 积分，迭代期望 vs Itô 引理）。
* **时间管理:** 4小时做4-5个大题，平均每题45-60分钟，需要快速定位问题核心，高效推导和计算。
