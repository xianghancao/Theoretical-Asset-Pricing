这个公式：

$$\mathbf{P_t = E_t\left[\sum^T_{s=t+1}D_s\frac{\zeta_{s}}{\zeta_t}\right]}$$

以及我们在上一个回答中使用的**迭代（或单期）**定价公式：

$$\mathbf{P_t = E_t [\zeta_{t+1} (D_{t+1} + P_{t+1})]}$$

**它们实际上是兼容的，只是表达形式不同。** 第一个公式是**多期（Multi-Period）**的**展开式**，它直接将所有未来的现金流折现回 $t$ 时刻。第二个公式是**动态（Dynamic）**的**递推式**，它体现了价格的动态一致性。

下面是它们兼容性的详细解释和推导。

---

## 动态一致性与多期展开式的兼容性

### 1. 理解多期展开式 (公式一)

$$\mathbf{P_t = E_t\left[\sum^T_{s=t+1}D_s\frac{\zeta_{s}}{\zeta_t}\right]}$$

这个公式的含义是：**资产在 $t$ 时刻的价格 $P_t$，等于所有未来股息 $D_s$ 的现值之和。**

* **折现因子 $\frac{\zeta_{s}}{\zeta_t}$：** 这是一个**累积状态价格平减指数**。它表示从 $t$ 时刻到 $s$ 时刻的**状态价格折现率**。在 $t=0$ 时刻，$\zeta_t = \zeta_0 = 1$（通常做此设定），则折现因子简化为 $\zeta_s$，即 $s$ 时刻的状态价格。

### 2. 理解单期递推式 (公式二)

$$\mathbf{P_t = E_t [\zeta_{t+1} (D_{t+1} + P_{t+1})]}$$

这个公式的含义是：**资产在 $t$ 时刻的价格 $P_t$，等于下一期（$t+1$）的总收益（股息 $D_{t+1}$ 加上下一期价格 $P_{t+1}$）的期望现值。** 这里的 $\zeta_{t+1}$ 是从 $t$ 到 $t+1$ 的**单期**状态价格平减指数。

### 3. 从递推式推导出展开式（证明兼容性）

我们可以通过**迭代**单期递推式来证明它等价于多期展开式。我们假设终点是 $T=2$（对应您的习题），且 $\zeta_0 = 1$。

#### **第一步： $P_{T-1}$ 的计算（$t=1$）**

终点价格 $P_T = P_2 = 0$。

$$P_1 = E_1 [\zeta_2 (D_2 + P_2)] = E_1 [\zeta_2 D_2]$$

#### **第二步： $P_{T-2}$ 的计算（$t=0$）**

将 $P_1$ 代入 $P_0$ 的递推公式：

$$P_0 = E_0 [\zeta_1 (D_1 + P_1)]$$

$$P_0 = E_0 \left[ \zeta_1 D_1 + \zeta_1 P_1 \right]$$

将 $P_1$ 替换为第一步的结果 $E_1 [\zeta_2 D_2]$：

$$P_0 = E_0 \left[ \zeta_1 D_1 + \zeta_1 \cdot E_1 [\zeta_2 D_2] \right]$$

根据**期望的迭代性质（Law of Iterated Expectations）**，对于任何随机变量 $X$ 和 $\zeta_1$，都有：

$$E_0 [\zeta_1 \cdot E_1 [X]] = E_0 [\zeta_1 X]$$

因此，我们可以写成：

$$E_0 \left[ \zeta_1 \cdot E_1 [\zeta_2 D_2] \right] = E_0 [\zeta_1 \zeta_2 D_2]$$

#### **最终结果：**

$$P_0 = E_0 [\zeta_1 D_1] + E_0 [\zeta_1 \zeta_2 D_2]$$

由于 $\zeta_0 = 1$，这恰好就是**多期展开式**在 $T=2, t=0$ 时的形式：

$$P_0 = E_0\left[\sum^2_{s=1}D_s\frac{\zeta_{s}}{\zeta_0}\right] = E_0[\zeta_1 D_1] + E_0[\zeta_2 D_2]$$

### 总结

这两个公式是**等价且完全兼容**的，它们只是描述了同一件事的不同方式：

* **递推式**更适合**计算和理解动态对冲**，因为它强制您**一步一步**地进行折现（先算 $P_2 \to P_1$，再算 $P_1 \to P_0$）。
* **展开式**更适合**概念和理论分析**，因为它将最终价格表达为所有未来现金流的**静态总和**。

在实际应用中，处理复杂的多期问题时，我们通常会使用**递推式**（即公式二）从后往前进行**反向归纳（Backward Induction）**，这和您第一问中的计算方法是完全一致的。
