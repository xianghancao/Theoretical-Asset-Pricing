## Core Textbooks

- Asset Pricing by John Cochrane: A standard PhD-level text covering theory and empirical methods.
- Financial Decisions and Markets by John Campbell: Focuses on modern asset pricing theory.
- Dynamic Asset Pricing Theory by Darrell Duffie: Advanced treatment of continuous-time models.
- Empirical Asset Pricing by Bali, Engle, and Murray: For empirical techniques and data analysis.

- 主流金融教材（例如Bodie、Kane和Marcus（2014年）、Brealey、Myers和Allen（2014年）或Hillier、Grinblatt和Titman（2011年））
