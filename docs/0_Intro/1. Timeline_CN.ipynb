# Timeline

## 现代投资组合理论（MPT）

- Markowitz (1952)：

    - 提出投资组合多元化降低风险（以标准差衡量），不改变预期回报。
    - 强调资产对投资组合整体风险的贡献，而非其自身波动。
    - 引入有效前沿（均值-标准差空间中的双曲线），表示最低方差的投资组合。


- Tobin (1958)：

    - 扩展Markowitz模型，加入无风险资产，提出两基金分离定理。
    - 投资者应投资于切点组合（风险资产组合）与无风险资产，比例根据风险厌恶程度调整。
    - 均值-方差选择仅在二次效用函数或正态分布回报条件下对期望效用最大化者最优，这两个假设均不现实。

## 资本资产定价模型（CAPM）

- Sharpe (1964)：

- 基于Markowitz和Tobin模型，推导CAPM，公式为：
  
$$E[r_i] = r_f + \beta_i (E[r_m] - r_f)$$

- 其中 $ \beta_i = \frac{\text{Cov}(r_i, r_m)}{\text{Var}(r_m)} $，衡量资产对市场组合系统性风险的贡献。
- 假设所有投资者按Tobin分离定理投资，切点组合等于市场组合。


- 其他贡献者：Lintner (1965)、Mossin (1966)、Treynor (1962) 也独立推导CAPM。
- CAPM的教学与局限：
    
    - CAPM是MBA金融课程核心，适合理解风险与回报关系。
    - 但其假设（二次效用或正态分布回报）不现实，且为单期模型，无法处理动态回报（如可预测性）。

## CAPM的实证失败

- 早期测试（1960-1980）：
    - 早期研究支持CAPM，高 $\beta$ 组合有较高回报，但测试资产组合可能不够有力。
 

- 现代测试（1980s后）：
    - Banz (1981)：小市值股票回报高于CAPM预测。
    - Fama and French (1992)：高账面市值比（价值股）回报高于CAPM预测。
    - 测试数据（1960-2020，25个按市值和账面市值比排序的组合）：
        - 价值股回报高于成长股（价值溢价），但β较低，违背CAPM预测。
        - 横截面回归显示β与回报关系弱（R²仅16%），时间序列回归α显著，拒绝CAPM。
     

- Roll (1977)批判：
    - CAPM唯一可测试假设是市场组合位于均值-方差有效前沿。
    - 若市场组合代理（如CRSP价值加权指数）非有效，α不为零，导致CAPM失败。
 

- 失败原因：
    - 代理市场组合可能非“真实”市场组合。
    - 单期模型无法捕捉动态风险溢价或多因子风险。


## CAPM失败的影响

- 基金表现评估：

    - 使用CAPM评估价值策略基金可能误判为“超常表现”，实为风格效应。


- 资本预算：

    - 对价值公司使用CAPM低估折现率，可能导致接受负NPV项目。


- 理论问题：

    - CAPM假设过于严格（二次效用或正态分布）。
    - 单期模型无法处理可预测回报或投资机会集变化。
    - 无法解释市场组合回报水平及其决定因素。






## 多因子模型

- 原因：CAPM失败表明存在其他定价风险源，多因子模型尝试捕捉这些风险。
- 因子类别：

    - 市场组合：通常包含在多因子模型中。
    - 反映投资者边际效用：如经济状态因子，投资者更重视经济低迷时高回报的资产（对冲衰退风险）。
    - 预测投资机会集变化：如预测市场回报的因子。
    - 捕捉股票回报共同变化：因子需反映系统性风险。


- Fama-French三因子模型 (1993)：

    - 加入SMB（小市值减大市值）和HML（高账面市值比减低账面市值比）因子。
    - 模型：
    $$r_{i,t} - r_{f,t} = \alpha_i + \beta_{i,m}(r_{m,t} - r_{f,t}) + \beta_{i,SMB}SMB_t + \beta_{i,HML}HML_t + \epsilon_{i,t}$$
    
    - 实证结果：三因子模型解释73%回报变异，优于CAPM的16%，但缺乏坚实理论基础（SMB和HML代理何种风险？）。
    





## 跨期资本资产定价模型（ICAPM）

- Merton (1973)：

    - 若股票回报可预测，投资机会集随时间变化，CAPM失效。
    - ICAPM引入状态变量（预测投资机会集变化的变量），如股息收益率、期限利差、违约利差等。


- 可预测性：

    - 1980s前认为股票回报不可预测，ICAPM应用有限。
    - 1980s后，Fama和French (1988, 1989)等发现股息收益率等变量可预测长期回报（R²达60%）。
    - 可预测性与市场有效性一致，若风险溢价随时间变化。


- 含义：

    - 投资者偏好对冲再投资风险的股票（在预测回报低时表现好），推高其价格，降低预期回报。
    





## 均值-方差框架在多因子环境下的变化

- 单一风险（市场风险）：

    - 投资者选择均值-方差有效组合（Markowitz-Tobin框架）。
    - 若存在无风险资产，投资于切点组合和无风险资产（两基金分离）。


- 多因子风险（如衰退风险）：

    - 投资者需对冲其他系统性风险（如衰退因子），优化问题受约束。
    - 优化组合不再均值-方差有效，而是多因子有效（在给定预期回报和因子敏感度下最小化方差）。
    - N基金分离定理：投资于无风险资产、市场组合及对冲其他风险的组合，权重依投资者偏好而定。


- 启示：

    - Sharpe比率（仅考虑均值和方差）可能不适合评估多因子世界中的投资组合表现。
