# 5. Implications of CAPM’s Failure

Fund Performance Evaluation:

A value strategy fund may show significant CAPM alphas, mistaken for superior performance, when it’s merely a style effect.


Capital Budgeting:

Using CAPM for value firms underestimates discount rates, potentially leading to accepting negative-NPV projects.


Theoretical Issues:

Restrictive assumptions (quadratic utility or normal returns).
Single-period model cannot handle predictable returns or changes in investment opportunities.
Does not explain the level or determinants of market portfolio returns.




## 6. Multifactor Models

Rationale: CAPM’s failure suggests other priced risk sources; multifactor models aim to capture these.
Factor Categories:

Market Portfolio: Almost always included.
Investor Marginal Utility: Factors reflecting economic states (e.g., recession risk). Assets with high payoffs in bad economic states are more valuable, commanding higher prices and lower expected returns.
Predicting Investment Opportunities: Factors forecasting changes in the investment opportunity set.
Common Variation: Factors must capture systematic risk affecting stock returns.


Fama-French Three-Factor Model (1993):

Adds SMB (small-minus-big) and HML (high-minus-low book-to-market) factors to the market factor.
Model:
$$r_{i,t} - r_{f,t} = \alpha_i + \beta_{i,m}(r_{m,t} - r_{f,t}) + \beta_{i,SMB}SMB_t + \beta_{i,HML}HML_t + \epsilon_{i,t}$$

Empirical Results: Explains 73% of return variation (vs. CAPM’s 16%), but lacks a solid theoretical foundation for SMB and HML (what risks do they proxy?).




## 7. Intertemporal Capital Asset Pricing Model (ICAPM)

Merton (1973):

If stock returns are predictable, the investment opportunity set varies over time, invalidating CAPM.
ICAPM includes state variables (e.g., dividend yield, term spread, default spread) that forecast changes in investment opportunities.


Predictability:

Pre-1980s, stock returns were considered unpredictable, limiting ICAPM’s application.
Post-1980s, Fama and French (1988, 1989) and others found evidence of predictability (e.g., dividend yield predicts 60% of 5-year return variation).
Predictability aligns with market efficiency if risk premiums are time-varying.


Implications:

Investors prefer stocks that hedge reinvestment risk (perform well when expected returns are low), increasing their prices and lowering expected returns.




## 8. Mean-Variance Framework in a Multifactor World

Single Risk (Market Risk):

Investors choose mean-variance efficient portfolios (Markowitz-Tobin framework).
With a risk-free asset, wealth is allocated between the tangency portfolio and the risk-free asset (two-fund separation).


Multifactor Risks (e.g., Recession Risk):

Investors hedge additional systematic risks (e.g., recession factor), making the optimization problem constrained.
Optimal portfolios are multifactor efficient (minimize variance for given expected return and factor sensitivity), not mean-variance efficient.
N-Fund Separation Theorem: Wealth is allocated among the risk-free asset, market portfolio, and portfolios hedging other risks, with weights varying by investor preferences.


Insight:

The Sharpe ratio (mean-variance based) may not be suitable for evaluating portfolios in a multifactor world.
