# Appendix: Tpyes of distribution of asset returns
How many types of distribution of asset returns can be?

The distribution of asset returns can take various forms, depending on the characteristics of the asset, market conditions, and the time horizon considered. In asset pricing and financial modeling, several types of distributions are commonly used or observed to describe asset returns. Below is a comprehensive overview of the main types of distributions that asset returns can follow, particularly in the context of a PhD-level asset pricing course:
## 1. Normal Distribution (Gaussian Distribution)

Description: The normal distribution assumes that asset returns are symmetrically distributed around the mean, with a bell-shaped curve. It is characterized by two parameters: mean ($\mu$) and standard deviation ($\sigma$).
Use in Asset Pricing:

Widely used in classical models like the Capital Asset Pricing Model (CAPM) and Black-Scholes option pricing due to its mathematical simplicity.
Assumes that returns are driven by many small, independent random shocks (Central Limit Theorem).


Characteristics:

Symmetric, with thin tails (low probability of extreme events).
Fully described by mean and variance.


Limitations:

Empirical evidence shows asset returns often exhibit fat tails (kurtosis) and skewness, which the normal distribution cannot capture.


Example: Daily stock returns are often approximated as normal in simple models, though this is a rough approximation.

## 2. Log-Normal Distribution

Description: If the logarithm of asset prices follows a normal distribution, the asset prices (and thus returns over discrete periods) follow a log-normal distribution. This ensures prices remain positive.
Use in Asset Pricing:

Common in models like Black-Scholes, where asset prices follow a geometric Brownian motion, implying log-returns are normally distributed.


Characteristics:

Skewed to the right (positive skewness).
Prevents negative asset prices, unlike the normal distribution.


Limitations:

Still assumes relatively thin tails, which may not capture extreme market events.


Example: Used to model stock prices over longer horizons in continuous-time finance.

## 3. Student’s t-Distribution

Description: The t-distribution has heavier tails than the normal distribution, making it suitable for modeling asset returns with a higher likelihood of extreme events (fat tails).
Use in Asset Pricing:

Often used in empirical asset pricing to model returns with high kurtosis.
Common in Bayesian finance models or robust statistical methods.


Characteristics:

Controlled by degrees of freedom ($v$): lower $v$ means fatter tails.
Symmetric but with higher kurtosis than the normal distribution.


Example: Used to model daily or high-frequency returns of stocks or portfolios, especially during volatile periods.

## 4. Skewed Distributions (e.g., Skew-Normal or Skew-t)

Description: These distributions allow for asymmetry in asset returns, capturing positive or negative skewness observed in financial data.
Use in Asset Pricing:

Useful for modeling returns of assets like options, equities, or hedge funds, where returns are often skewed (e.g., negative skewness for equities due to crashes).


Characteristics:

Can model asymmetry in addition to fat tails.
Examples include the skew-normal distribution or skewed versions of the t-distribution.


Example: Equity index returns often exhibit negative skewness, reflecting larger downside risks.

## 5. Stable Distributions (e.g., Lévy-Stable)

Description: Stable distributions generalize the normal distribution and allow for heavy tails and skewness. They are characterized by four parameters: stability ($\alpha$), skewness ($\beta$), scale, and location.
Use in Asset Pricing:

Used to model asset returns with extreme events, as they allow for infinite variance (when $\alpha < 2$).
Relevant in high-frequency data or markets with jumps.


Characteristics:

Heavy tails (more extreme events than normal).
Includes normal distribution as a special case ($\alpha = 2$).


Limitations:

Difficult to estimate due to complex parameterization.


Example: Used to model cryptocurrency returns or high-frequency trading data.

## 6. Mixture Distributions (e.g., Gaussian Mixture Models)

Description: A mixture distribution combines multiple distributions (e.g., two or more normal distributions) to model returns with complex behavior, such as regime-switching or clustering.
Use in Asset Pricing:

Captures regime-dependent behavior (e.g., bull vs. bear markets).
Used in volatility modeling or to account for different market states.


Characteristics:

Can model skewness, kurtosis, and multimodality.
Flexible but requires estimation of multiple parameters.


Example: Used to model stock returns during normal vs. crisis periods.

## 7. Generalized Hyperbolic Distributions

Description: A flexible class of distributions that includes normal, t, and other distributions as special cases. It can model skewness and heavy tails.
Use in Asset Pricing:

Popular in empirical finance for modeling asset returns with complex tail behavior.
Used in risk management and option pricing.


Characteristics:

Highly flexible, with parameters for location, scale, skewness, and tail heaviness.
Computationally intensive to estimate.


Example: Applied to model returns of hedge funds or commodities.

## 8. Nonparametric or Empirical Distributions

Description: Instead of assuming a specific functional form, nonparametric methods use the empirical distribution of historical returns or kernel density estimation.
Use in Asset Pricing:

Useful when the true distribution is unknown or does not fit standard parametric forms.
Common in simulation-based methods like Monte Carlo or bootstrap.


Characteristics:

No assumptions about shape, allowing for maximum flexibility.
Data-driven but requires large datasets for accuracy.


Example: Used in stress testing or historical simulation for portfolio risk.

## 9. Jump-Diffusion and Lévy Processes

Description: These distributions model asset returns as a combination of continuous diffusion (e.g., Brownian motion) and discrete jumps, capturing sudden price movements.
Use in Asset Pricing:

Common in option pricing (e.g., Merton’s jump-diffusion model) and modeling high-frequency returns.
Captures extreme events like market crashes or news-driven spikes.


Characteristics:

Combines normal-like diffusion with Poisson-driven jumps.
Allows for fat tails and skewness.


Example: Used to model returns of stocks during earnings announcements or market crises.

## 10. Extreme Value Distributions

Description: Focuses on the tails of the return distribution, modeling extreme events using distributions like the Generalized Extreme Value (GEV) or Generalized Pareto Distribution (GPD).
Use in Asset Pricing:

Used in risk management (e.g., Value-at-Risk, Expected Shortfall) to model rare but impactful events.


Characteristics:

Specifically designed for tail behavior.
Asymmetric and heavy-tailed.


Example: Applied to study tail risks in equity or bond markets.

Key Considerations in Asset Pricing

Empirical Evidence: Real-world asset returns often deviate from normality, exhibiting:

Fat tails (leptokurtosis): Higher probability of extreme returns.
Skewness: Negative skewness is common in equity returns due to downside risks.
Time-varying volatility: Returns often show clustering (e.g., modeled by GARCH models).


Model Choice: The choice of distribution depends on the asset class (e.g., equities, bonds, options), time horizon (daily vs. monthly), and purpose (e.g., pricing, risk management, or portfolio optimization).
PhD-Level Focus: At the PhD level, you’ll likely encounter:

Theoretical discussions of these distributions in stochastic processes (e.g., Brownian motion, Lévy processes).
Empirical testing using likelihood-based methods, GMM, or Bayesian approaches to estimate distribution parameters.
Coding assignments to simulate or fit these distributions (e.g., using Python or R).



Practical Steps for Studying Distributions in Asset Pricing

Learn Theoretical Foundations:

Study stochastic processes (e.g., Stochastic Calculus for Finance by Shreve).
Understand how distributions arise in models like CAPM, Black-Scholes, or multifactor models.


Empirical Analysis:

Use datasets (e.g., CRSP, Compustat) to test which distributions fit historical returns.
Practice fitting distributions using maximum likelihood or moment-based methods.


Coding:

Simulate returns using Python (e.g., numpy, scipy) or R for different distributions.
Example: Simulate log-normal returns for a stock price path or fit a t-distribution to daily returns.


Read Key Papers:

Mandelbrot (1963) on stable distributions.
Fama and French (1993) on empirical deviations from normality.
Cont (2001) on stylized facts of financial returns.
