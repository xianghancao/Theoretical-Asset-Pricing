# Prerequisite: Risk Aversion

## Absolute Risk Aversion (ARA)

### 绝对风险规避的定义

公式：$ R_A(Y) := -\frac{u''(Y)}{u'(Y)} $

这里，$ u(Y) $ 是效用函数，$ u'(Y) $ 是它的第一导数（表示财富增加时效用的变化率），$ u''(Y) $ 是第二导数（表示效用变化的加速或减缓）。
$ R_A(Y) $ 是一个度量，告诉我们一个人对风险的绝对厌恶程度如何随财富 $ Y $ 变化。


### 直观理解：

想象你是一个投资者，$ R_A(Y) $ 衡量你愿意为避免风险（比如投资不确定的资产）而放弃多少确定收益。
这个公式通过效用函数的曲率（第二导数）来计算：如果效用函数弯曲得很厉害（$ u''(Y) $ 很大且负），说明你对风险很敏感，风险规避程度高。




### 为什么叫“绝对风险规避”？

“绝对”指的是这种风险规避与财富的绝对数量（而不是比例）有关。比如，失去 1000 欧元对你的影响不取决于你有多少钱，而是取决于这 1000 欧元本身的价值。
对比“相对风险规避”（后面会提到），后者关注财富的百分比变化。


假设递减的绝对风险规避 (DARA)

意思：随着财富增加，一个人愿意投资于风险资产的金额会增加。这是因为他们变得不太害怕风险。
例子：

假设一个人的财富是 20,000 欧元，他只愿意投资 2000 欧元在风险资产（比如股票）。
当他的财富增加到 50,000 欧元时，他愿意投资 5000 欧元在风险资产。
这种行为表明他的绝对风险规避在减少（DARA）：财富越多，他愿意承担的风险越多。


### 为什么合理？

直觉上，穷人失去 1000 欧元会感觉很痛苦（风险规避高），而富人失去 1000 欧元影响较小（风险规避低）。DARA 反映了这种心理。




### 难点解析

公式不理解？

$ u''(Y) $ 是效用函数的“凹度”。如果 $ u''(Y) < 0 $（大多数效用函数都是凹函数），说明效用随财富增加而减慢，风险规避存在。
$ u'(Y) $ 是边际效用，衡量额外的财富带来的满足感。
$ R_A(Y) $ 的负号确保风险规避是正值，因为 $ u''(Y) $ 通常是负的。


例子不明白？

想象你有 20,000 欧元，只敢放 2000 欧元在股票（因为怕亏）。但当你有 50,000 欧元时，你可能愿意放 5000 欧元，因为多出来的 30,000 欧元让你更有安全感，愿意多冒险。
这说明你的风险规避随财富增加而下降，符合 DARA。

## Relative Risk Aversion (RRA)

### 相对风险规避的定义

公式：$ R_R(Y) := Y \cdot R_A(Y) = -Y \frac{u''(Y)}{u'(Y)} $

其中：

$ R_A(Y) = -\frac{u''(Y)}{u'(Y)} $ 是你之前看到的绝对风险规避。
$ u(Y) $ 是效用函数，$ u'(Y) $ 是第一导数（财富增加带来的效用变化），$ u''(Y) $ 是第二导数（效用变化的快慢）。
$ Y $ 是财富的总量。


意思：$ R_R(Y) $ 衡量一个人愿意把财富的多少比例投入风险资产（比如股票），而这个比例与财富的绝对值相关。


### 直观理解：

相对风险规避关注的是你愿意拿财富的“百分比”去冒险，而不是固定金额。
比如，失去 10% 财富的影响对你来说可能比失去 1000 欧元的绝对金额更重要。




### 为什么叫“相对风险规避”？

“相对”指的是这种风险规避是基于财富的比例（百分比），而不是像绝对风险规避那样只看固定的金额。
它告诉你，随着财富增加，你愿意冒险的比例是否变化。


假设恒定相对风险规避 (CRRA)

意思：如果一个人的相对风险规避是恒定的（不随财富变化），他们愿意投资于风险资产的财富比例保持不变。
例子：

假设一个人有 20,000 欧元财富，愿意投资 10%（即 2000 欧元）在风险资产。
当财富增加到 50,000 欧元时，他仍然愿意投资 10%（即 5000 欧元）在风险资产。
这说明他的相对风险规避是恒定的（CRRA）：他总是愿意拿相同比例的财富去冒险。


### 为什么合理？

有些人可能觉得风险应该按财富比例来考虑。比如，富人愿意拿 10% 财富冒险，穷人也愿意拿 10% 冒险，因为这对他们来说是“相对”可接受的损失。




### 难点解析

公式不理解？

$ u''(Y) $ 决定效用函数的凹度。如果 $ u''(Y) < 0 $（大多数效用函数是凹的），说明你越富，额外的财富带来的满足感越少，风险规避存在。
$ u'(Y) $ 是边际效用，衡量财富增加的“价值”。
$ Y \cdot R_A(Y) $ 把绝对风险规避乘以财富，变成一个相对的度量（比例）。


例子不明白？

想象你有 20,000 欧元，敢拿 2000 欧元（10%）买股票，因为你觉得这风险可控。
后来你有 50,000 欧元，你还是敢拿 5000 欧元（10%）买股票，而不是固定拿 2000 欧元。
这表明你的风险规避是“相对”的，比例没变，符合 CRRA。
