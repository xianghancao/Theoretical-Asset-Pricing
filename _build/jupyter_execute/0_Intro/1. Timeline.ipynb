# Timeline

## 1. Course Objectives and Structure

Objective: Provide a concise introduction to fundamental topics in theoretical asset pricing, emphasizing intuitive understanding over mathematical rigor (reserved for Part 2). The goal is to familiarize students with the stochastic discount factor (SDF) representation, i.e., $ P = E[zD] $ or $ 1 = E[zR] $, where $ z $ is the stochastic discount factor.

Structure: Covers asset pricing overview, utility theory, investment decisions, mean-variance analysis, CAPM, Arrow-Debreu securities, market completeness, SDF, fundamental theorem of asset pricing, consumption-based asset pricing, equity premium puzzle, and an introduction to continuous-time stochastic processes.


## 2. Modern Portfolio Theory (MPT)

Markowitz (1952):

Introduced diversification to reduce risk (measured as standard deviation) without altering expected portfolio returns.
Emphasized an asset’s contribution to the portfolio’s overall risk, not its standalone volatility.
Developed the efficient frontier (a hyperbola in mean-standard deviation space), representing portfolios with minimum variance for each expected return level.


Tobin (1958):

Extended Markowitz’s model by including a risk-free asset, leading to the two-fund separation theorem.
Investors should allocate wealth between the tangency portfolio (risky assets) and the risk-free asset, with proportions based on risk aversion.
Mean-variance optimization is optimal for expected utility maximizers only under restrictive conditions: quadratic utility or normally distributed returns, both of which are unrealistic.




3. Capital Asset Pricing Model (CAPM)

Sharpe (1964):

Derived CAPM based on Markowitz and Tobin’s models, with the formula:
$$E[r_i] = r_f + \beta_i (E[r_m] - r_f)$$
where $ \beta_i = \frac{\text{Cov}(r_i, r_m)}{\text{Var}(r_m)} $, measuring an asset’s systematic risk relative to the market portfolio.
Assumes all investors follow Tobin’s separation theorem, making the tangency portfolio equal to the market portfolio.


Other Contributors: Lintner (1965), Mossin (1966), and Treynor (1962) also derived CAPM independently.
CAPM’s Role and Limitations:

Widely taught in MBA finance courses as an introduction to risk and return.
Relies on strong assumptions (quadratic utility or normal returns) and is a single-period model, unable to handle dynamic return patterns (e.g., predictability).




4. Empirical Failure of CAPM

Early Tests (1960–1980):

Supported CAPM, showing higher-beta portfolios had higher returns, but test portfolios may not have been robust.


Modern Tests (Post-1980s):

Banz (1981): Small-cap stocks yielded higher returns than CAPM predicted.
Fama and French (1992): Value stocks (high book-to-market ratios) outperformed CAPM predictions.
Test Data (1960–2020, 25 portfolios sorted by size and book-to-market):

Value premium: Value stocks had higher returns but lower betas, contradicting CAPM.
Cross-sectional regression showed a weak relationship between beta and returns (R² = 16%).
Time-series regressions produced significant alphas, rejecting CAPM.




Roll’s (1977) Critique:

CAPM’s only testable implication is that the market portfolio is mean-variance efficient.
If the proxy (e.g., CRSP value-weighted index) is not efficient, alphas are non-zero, leading to CAPM’s failure.


Reasons for Failure:

The proxy market portfolio may not represent the “true” market portfolio.
Single-period model cannot capture dynamic risk premiums or multiple risk factors.




5. Implications of CAPM’s Failure

Fund Performance Evaluation:

A value strategy fund may show significant CAPM alphas, mistaken for superior performance, when it’s merely a style effect.


Capital Budgeting:

Using CAPM for value firms underestimates discount rates, potentially leading to accepting negative-NPV projects.


Theoretical Issues:

Restrictive assumptions (quadratic utility or normal returns).
Single-period model cannot handle predictable returns or changes in investment opportunities.
Does not explain the level or determinants of market portfolio returns.




6. Multifactor Models

Rationale: CAPM’s failure suggests other priced risk sources; multifactor models aim to capture these.
Factor Categories:

Market Portfolio: Almost always included.
Investor Marginal Utility: Factors reflecting economic states (e.g., recession risk). Assets with high payoffs in bad economic states are more valuable, commanding higher prices and lower expected returns.
Predicting Investment Opportunities: Factors forecasting changes in the investment opportunity set.
Common Variation: Factors must capture systematic risk affecting stock returns.


Fama-French Three-Factor Model (1993):

Adds SMB (small-minus-big) and HML (high-minus-low book-to-market) factors to the market factor.
Model:
$$r_{i,t} - r_{f,t} = \alpha_i + \beta_{i,m}(r_{m,t} - r_{f,t}) + \beta_{i,SMB}SMB_t + \beta_{i,HML}HML_t + \epsilon_{i,t}$$

Empirical Results: Explains 73% of return variation (vs. CAPM’s 16%), but lacks a solid theoretical foundation for SMB and HML (what risks do they proxy?).




7. Intertemporal Capital Asset Pricing Model (ICAPM)

Merton (1973):

If stock returns are predictable, the investment opportunity set varies over time, invalidating CAPM.
ICAPM includes state variables (e.g., dividend yield, term spread, default spread) that forecast changes in investment opportunities.


Predictability:

Pre-1980s, stock returns were considered unpredictable, limiting ICAPM’s application.
Post-1980s, Fama and French (1988, 1989) and others found evidence of predictability (e.g., dividend yield predicts 60% of 5-year return variation).
Predictability aligns with market efficiency if risk premiums are time-varying.


Implications:

Investors prefer stocks that hedge reinvestment risk (perform well when expected returns are low), increasing their prices and lowering expected returns.




8. Mean-Variance Framework in a Multifactor World

Single Risk (Market Risk):

Investors choose mean-variance efficient portfolios (Markowitz-Tobin framework).
With a risk-free asset, wealth is allocated between the tangency portfolio and the risk-free asset (two-fund separation).


Multifactor Risks (e.g., Recession Risk):

Investors hedge additional systematic risks (e.g., recession factor), making the optimization problem constrained.
Optimal portfolios are multifactor efficient (minimize variance for given expected return and factor sensitivity), not mean-variance efficient.
N-Fund Separation Theorem: Wealth is allocated among the risk-free asset, market portfolio, and portfolios hedging other risks, with weights varying by investor preferences.


Insight:

The Sharpe ratio (mean-variance based) may not be suitable for evaluating portfolios in a multifactor world.




9. Other Topics (Not Expanded)

Utility Theory and Risk Aversion: How investors balance risk and return.
Arrow-Debreu Securities and Market Completeness: Concepts of state prices and complete markets.
SDF Representation: Core asset pricing framework, with existence theorems based on no-arbitrage and law of one price.
Consumption-Based Asset Pricing and Equity Premium Puzzle: Challenges in consumption-based models.
Continuous-Time Stochastic Processes and Itô’s Lemma: Foundation for mathematical derivations.


###families: Key Takeaways

Core Frameworks: From Markowitz’s diversification to Tobin’s separation theorem and Sharpe’s CAPM, asset pricing theory evolved, but CAPM is limited by strict assumptions and empirical failures.
Empirical Challenges: Value and size premiums highlight CAPM’s inability to explain return variations, necessitating multifactor models like Fama-French.
Theoretical Advances: ICAPM and multifactor models address dynamic returns and multiple risks, emphasizing hedging economic state changes.
Practical Implications: CAPM’s limitations affect fund evaluation and capital budgeting, requiring more sophisticated models.
